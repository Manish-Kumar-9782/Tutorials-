## Entry Widget config

```python
class Entry_widget():
    
    def __init__(self):
        # h = heighlight
        self.h_background = 'black'
        self.h_color = 'black'
        self.h_thickness = 1
    
    def highlight_border(self, EntryWidget,color=None):
        if color==None:    
            self.h_background = 'black'
            self.h_color = 'black'
        else:
            self.h_background = color
            self.h_color = color
            
        EntryWidget.config(highlightbackground = self.h_background)
        EntryWidget.config(highlightcolor = self.h_color)
        EntryWidget.config(highlightthickness = self.h_thickness)
    
    def default_border(self,EntryWidget):
        self.h_background = 'black'
        self.h_color = 'black'
        EntryWidget.config(highlightbackground = self.h_background)
        EntryWidget.config(highlightcolor = self.h_color)
        EntryWidget.config(highlightthickness = self.h_thickness)
        
   
```

```python
class Write():
    
    def __init__(self, path=None):
        
        if path == None:
            self.path = "F:/Projects Data/Password manager/Users/"
            
        else:
            self.path = path
            
        self.filename ="Config.txt"
        
        
    def parse_text(self,Data):
        """
        This method will create a new config file for the current user.
        Data is a list or dict of User detalis.
        """
        filepath  = os.path.join(self.path,Data["User Name"])
        # first check dir already exist
        
        if os.path.isdir(filepath)!=True:
            os.mkdir(filepath)
            
            file = open(os.path.join(filepath, self.filename), 'w')
        
            for key in Data:
                file.write(key+"="+Data[key]+'\n')

            file.close()
        else:
            user = Data["User Name"] 
            msg = f"User '{user}' is already exist try different User."
            messagebox.showerror(title="Regestration Error..", message=msg)
            
    
            
```

1. Get access the parent window and use `parent.winfo_childern()` it will return all the widget.

3. Now we have a list of all the widget and found the our desired widget label or Entry
4. Label or Entry both have `textvariable`option so we need to get their respective variable
5. For that we need to make a variable for every label or entry which we want access later.
6. We should define the variable at time of difining the Label and Entry widgets.
7. EVery Label of Entry widget Variable can be accessed from their optinal functions
8. So we will use `getvar()` and `setvar()` metod to get and set the variable values
9. for that we need to assign the name before accessing the variable
10. this is need to done at the defination of the variable like `var = StringVar(name="var1")`
