**Creating the button**

To create the button we will use the `tk.Button()` class from the tkinter module.

````python
# creating the main window

root = tk.Tk()

root.title("Button Tutorial")
root.geometry("400x300")
#----------------------------------------------------------------------------------------------------------------------

# Now here we will create the button but we will not add any thing to it , just a simple button which will do nothing

# to create the button we need a top level or a parent window and we have a parent window as the root window.
# Here we will use the tk.Button() class to create the button but we will not add any thing in it

button = tk.Button(root, text = "Button 1")
button.pack(pady = 20)

# we have have created the button but this button is not doing anyting until we do not tell him what to do
# this button is the blank slate in which we can write any thing, we perform any kind of the task by using this button.
# In next cell we will see about that.
#----------------------------------------------------------------------------------------------------------------------
root.mainloop()
````

**Perform A task with a button**

Now in this we will see how we can perform a task by using the button. To perfrom a task we will use the `command` option.

````python
# creating the main window

root = tk.Tk()

root.title("Button Tutorial")
root.geometry("400x300")
#----------------------------------------------------------------------------------------------------------------------

# In above cell we have seen that how we created the button and placed it on the root window 
# Now we will add some task in it, when it pressed that task will happen.

# for this we need to create a function or we can use another predefined funtion but for now we will use the user defined funtion.

# Now we will print a message on the window.
# we need a label to show the message.

Text = tk.StringVar()
label = tk.Label(root, textvariable = Text )
label.pack(pady = 20)

# this label will be placed top of the button becaues it is packed before the message 

Text.set("Hello")

def say_hi():
    
    """
        This fucntion will say hi to the user when the say button is pressed. This say message will be printed on
        the window in which the say button is placed.
        
        We will use the label which is defined above this function.
    """
    Text.set("Hi, Manish")
    # NOw we have set the new text to the label but this will not show until we dont tell the button to pick this function,
    
# now we need to use the command option in the Button options. 

button = tk.Button(root, text = "Button" , command = say_hi)
button.pack(pady = 20)
# if we pack the label here then the label message would be printed blelow the button.
# label.pack()


# Now when you will pressed the button the message will change.
# We can do any thing with this button but for now we stay with this.

# we will see the other operation of through the button when we will work on the projects.
#----------------------------------------------------------------------------------------------------------------------

root.mainloop() 
````

**Working with some other additional options**

Now we will work on the Some additional features of the button .

##### 1. Active background 
Active background is the background color when the button is active, means pressed.

activebackground  is used when we want to change the background color of the button widget and show some fancy color on the button.

````python
# creating the main window

root = tk.Tk()

root.title("Button Tutorial")
root.geometry("400x300")
#----------------------------------------------------------------------------------------------------------------------



Text = tk.StringVar()
label = tk.Label(root, textvariable = Text )
label.pack(pady = 20)

# this label will be placed top of the button becaues it is packed before the message 

Text.set("Hello")

def say_hi():
    
    """
        This fucntion will say hi to the user when the say button is pressed. This say message will be printed on
        the window in which the say button is placed.
        
        We will use the label which is defined above this function.
    """
    Text.set("Hi, Manish")
    # NOw we have set the new text to the label but this will not show until we dont tell the button to pick this function,
    
# we want to change the color of the button when it is pressed.
# for that we need to use the activebackground option and we need to put a color name as the value of the activebackground option.
button = tk.Button(root, text = "Button" , command = say_hi , activebackground = 'blue')
button.pack(pady = 20)

# Now when you pressed the button the color will change of the button.

#----------------------------------------------------------------------------------------------------------------------

root.mainloop() 
````

##### 2. Active foreground
Active foreground is also an option which is similar to the activebackground option but in this we changes the text color of the button.

Both `activebackground` and the `activeforeground` can be applied together. In this section we will use both activebackground and activeforground. 

````python
# creating the main window

root = tk.Tk()

root.title("Button Tutorial")
root.geometry("400x300")
#----------------------------------------------------------------------------------------------------------------------



Text = tk.StringVar()
label = tk.Label(root, textvariable = Text )
label.pack(pady = 20)

# this label will be placed top of the button becaues it is packed before the message 

Text.set("Hello")

def say_hi():
    
    """
        This fucntion will say hi to the user when the say button is pressed. This say message will be printed on
        the window in which the say button is placed.
        
        We will use the label which is defined above this function.
    """
    Text.set("Hi, Manish")
    # NOw we have set the new text to the label but this will not show until we dont tell the button to pick this function,
    
# Now we will use the activebackground and activeforeground property together.
    
button = tk.Button(root, text = "Button" , command = say_hi , activebackground = 'blue' , activeforeground = 'white')
button.pack(pady = 20)

# Now when you pressed the button the color will change of the button.
# Now the text color also would change after pressing the button.|

#----------------------------------------------------------------------------------------------------------------------

root.mainloop() 
````

#### 3. Disable the button

In this we will see: `disabledforeground`, `state=tk.DISABLED` 

`disabledforeground:` it is used when we want to show disabled thinks with some other color.

`state=tk.DISABLED:` it is used to disable the widget, in this case we will disable the button and make it unresponsive. 

````python
# creating the main window

root = tk.Tk()

root.title("Button Tutorial")
root.geometry("400x300")
#----------------------------------------------------------------------------------------------------------------------



Text = tk.StringVar()
label = tk.Label(root, textvariable = Text )
label.pack(pady = 20)

# this label will be placed top of the button becaues it is packed before the message 

Text.set("Hello")

def say_hi():
    
    """
        This fucntion will say hi to the user when the say button is pressed. This say message will be printed on
        the window in which the say button is placed.
        
        We will use the label which is defined above this function.
    """
    Text.set("Hi, Manish")
    # NOw we have set the new text to the label but this will not show until we dont tell the button to pick this function,
    
# Adding the state and disabledforeground options of the button.
    
button = tk.Button(root, text = "Button" , command = say_hi , disabledforeground = 'red', state = tk.DISABLED)
button.pack(pady = 20)

# Now the button is disabled and the text color is red to show that button is not available

#----------------------------------------------------------------------------------------------------------------------

root.mainloop() 
````

#### 5. Image

But we cant use put the Image in simple way fisrt we need to make the image object which is compataible with the tkinter and then put it to the image option.
To make an Image-Object we need to use the Python Image Library `PIL`. we need to use:
```ruby

```
Now we need ot use the `PhotoImage` function to create an Image-Object. For that first we will import the `ImageTk, Image` classes from the `PIL` module.

````python
from PIL import ImageTk , Image
````

Since we have imported the required Classes from the PIL module we can move ahead.

````python
# creating the main window

root = tk.Tk()

root.title("Button Tutorial")
root.geometry("400x300")
#----------------------------------------------------------------------------------------------------------------------

# Now we will make an Image-Object 
# I have stored the the images inside "./bitmap icons/" which is stored at the current location at which this program is saved.

# To create a Image-Object we need to things
# 1. PhotoImage() object from the ImageTk class
# 2. Address of the image

# creating the image object



image = ImageTk.PhotoImage(image=Image.open('bitmap icons/applications.png'))
# here we had to use the 'Image.open()' function to first initialized the image 
# Note: without initializing the image we cant use the image.
# Now we have made an Image-Object now we will use this object iside the Button widget.

Text = tk.StringVar()
label = tk.Label(root, textvariable = Text )
label.pack(pady = 20)

# this label will be placed top of the button becaues it is packed before the message 

Text.set("Hello")

def say_hi():
    
    """
        This fucntion will say hi to the user when the say button is pressed. This say message will be printed on
        the window in which the say button is placed.
        
        We will use the label which is defined above this function.
    """
    Text.set("Hi, Manish")
    # NOw we have set the new text to the label but this will not show until we dont tell the button to pick this function,
    
    
button = tk.Button(root, text = "Button" , command = say_hi , image = image)
# Note: if we are givening both text and the image option then it will ignore the text and only will display the image.

button.pack(pady = 20)

# Now we can see the image on the button but text will be disappeared.

#----------------------------------------------------------------------------------------------------------------------

root.mainloop() 
````

#### 6. overlelief
The relirf style to be used while the mouse is on the button,  default relief style is tk.RAISED or 'raised'

More styles:
* ridge
* raised
* groove
* flat
* sunken

````python
# creating the main window

root = tk.Tk()

root.title("Button Tutorial")
root.geometry("400x300")
#----------------------------------------------------------------------------------------------------------------------

# In this we will add the hovering effect on the button 
# button will change its relief style when we put the mouse button on it.

# Option is used: overrelief

Text = tk.StringVar()
label = tk.Label(root, textvariable = Text )
label.pack(pady = 20)

# this label will be placed top of the button becaues it is packed before the message 

Text.set("Hello")

def say_hi():
    
    """
        This fucntion will say hi to the user when the say button is pressed. This say message will be printed on
        the window in which the say button is placed.
        
        We will use the label which is defined above this function.
    """
    Text.set("Hi, Manish")
    # NOw we have set the new text to the label but this will not show until we dont tell the button to pick this function,
    
# Now to show the effect we need to use the relief and the overrelief option 
# relief option will define the its normal relief style and
# overrelief will change its relief style when the mouse is on it
    
button = tk.Button(root, text = "Button" , command = say_hi , overrelief = 'raised', relief = 'groove')
# Note: if we are giving the relief and the overrelief styles same then no effect would be visible

button.pack(pady = 20)

# Result: Butoon's Normal relief style is groove and it will become raised when the mouse is on the button.

#----------------------------------------------------------------------------------------------------------------------

root.mainloop() 
````

#### 7. underline
Default is -1 , meaning that no character of the text on the button will be underlined. if nonnegative, the corrsponding text character will be underlined.


````python
# creating the main window

root = tk.Tk()

root.title("Button Tutorial")
root.geometry("400x300")
#----------------------------------------------------------------------------------------------------------------------

# Current option:
    
    # Now we will make underline on character, in this only one character will be underlined , this will not underline the whole text 
    # Also this will not underline the texts more than one.


Text = tk.StringVar()
label = tk.Label(root, textvariable = Text )
label.pack(pady = 20)

# this label will be placed top of the button becaues it is packed before the message 

Text.set("Hello")

def say_hi():
    
    """
        This fucntion will say hi to the user when the say button is pressed. This say message will be printed on
        the window in which the say button is placed.
        
        We will use the label which is defined above this function.
    """
    Text.set("Hi, Manish")
    # NOw we have set the new text to the label but this will not show until we dont tell the button to pick this function,

# underline option only takes an integer as input.
    
button = tk.Button(root, text = "Button" , command = say_hi , underline = 0)
# Note: in this this will underline the first character of the given text
# if we pass underline = 3 then it will underline the 4th character of the text.


button.pack(pady = 20)

# Result: Now we will found that the charater 'B' of the 'Button' text's will be underlined.

#----------------------------------------------------------------------------------------------------------------------

root.mainloop() 
````

#### 8. wraplength
if this value is set to a positive number, the text lines will be wrqpped to fit within this length. for this we need at least 2-3 words on the button.

````python
# creating the main window

root = tk.Tk()

root.title("Button Tutorial")
root.geometry("400x300")
#----------------------------------------------------------------------------------------------------------------------

# Current option: wraplength
    
    # wraplength is the length of the text or characters in a single line.
    # if we specify this length then the characters which comes in that length range will be printed on the same line 
    # and other characters will shifted to the next line


# Now we will use wraplength option as 50 
    
button = tk.Button(root, text = "Press this Button" , command = say_hi , wraplength = 50)
# Note: In this we passed the 50 (interger) as the value of the length for the character to be displayed in single line.
# if we pass wraplength below 10 then it is possible that only one character will display in single line.
# because a single character can takes 10+ pixle to be displayed on the screen.


button.pack(pady = 20)

# Result: Words which comes in the given pixles range (wraplength = 50) will be displayed in the single line.

#----------------------------------------------------------------------------------------------------------------------

root.mainloop() 
````
