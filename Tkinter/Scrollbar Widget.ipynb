## Scrollbar 

The scrollbar widget is used to scroll down the content od the other widgets like listbox, text and  canvas, However, we can also create the horizontal scrollbars to the Entry widget.

```ruby
scrollbar = tk.Scrollbar(top|root|master , options....)
```

`Comman options:`
* activebackground
* bg or background
* bd or borderwidth
* command
* cursor
* highlightbackground
* highlightcolor
* hightlightthickness
* takefocus
* width

`Additional options:`
* elementborderwidth
* jump
* orient
* repeatdelay
* repeatinterval
* throughcolor

`Methods:`
* get(): It returnsthe two numbers a and b which represents the current position of the scrollbar.
* set(): It is used to connect the crollbar to the other widget w. The yscrollcommand or xscrollcommand of the other widget to this method.

#### Making a Scroball to scroll the Listbox

````python
# creating a scrollbar to scroll the Listbox.

root = tk.Tk()

root.title("Scrollbar Tutorial")
root.geometry("300x300")
root.iconbitmap("./bitmap icons/home.ico")
#------------------------------------------------------------------------------------------------

    # Now we will make a scrollbar Widget.
    # Since scrollbar doesnt work alone so we need one more widget which supports the scrolling 
    
    # making a scrollbar
scroll = tk.Scrollbar(root)
scroll.pack(side = tk.RIGHT, fill=tk.Y)

    # So, Now we wlil make a Listbox widget also which would be connected to the scrollbar
List = tk.Listbox(root , yscrollcommand = scroll.set)
    # here we need to use the x or yscrollcommand to set the scrollbar
    
    # now we need to insert some lines inside the Listbox so we can have the line out of the widget
for line in range(30):
    
    List.insert(tk.END,"Number"+str(line))
    
    # now we need to pack the Listbox 
List.pack(side = tk.LEFT)

# NOw we need to configure the scrollbar in which we  need to connect the command option to 
# the Listbox xview or yview accroding to the requirment.

scroll.config(command = List.yview)

    
    
"""
    Result: Here we have created a Listbox also and linked it to the scrollbar, 
            Here they are seperated with some space which make them to different widget, but in 
            real they are connected to each other.
            
    Note: We should always make a differnt Frame widget to make this kind of connecting object,
          The advantage of this arrangement is that they will be together, if the Frame is not large
          in size. And they will be isolated to the other widget in a complex program.
    
    Process: So the complete process is defined in the next Markdown cell.
        
        
"""  
#------------------------------------------------------------------------------------------------

root.mainloop()
````

#### Process to create a Proper Scrollbar

Since Scrollbar is not used alone , it has to be used to with anoter widget which supprot scrolling like Listbox, Canvas, Text etc. In this we need to connect each other, for that we need to follow some steps which are mentioned below.

1. Make a Frame with a proper defined Height and width in the root or a master window.
2. Make a scrollbar widget by using the Frame as the master window for the Scrollbar and place it to the desired position. (Assume we have made a Scrollbar named as SB)
3. Now make a widget `(master = Frame )`which supports the scrolling like Listbox, We will use the Listbox as an example.
4. Connect the Listbox to the Scrollbar (SB) by using any of two commands `(xscrollcommand or yscrollcommand)` and put the values as 'scrollbar.set' (SB.set) in which SB can be assumed the scrollbar and set is the scrollbar method to set it for the Lisbox in y dirction.
5. After packing the Listbox we also make to some configuration to the Scrollbar. Here in scrollbar we need to tell it that you are connected to this widget by using the command option, in our case we will give it as `(command = List.yview)`, xview or yview can be slected according to the orientation of the widget and dirction of the scrolling.


#### Making a Listbox-scrollbar in a Frame

`Note:`  We have already made a tutorial on how to make the Please check it if don't know about the Frame widget.

````python
## creating a scrollbar to scroll the Listbox in a single isolated Frame.

root = tk.Tk()

root.title("Scrollbar Tutorial")
root.geometry("300x300")
root.iconbitmap("./bitmap icons/home.ico")
#------------------------------------------------------------------------------------------------

    # Now first of all we need to make a Frame in which we will place the scrollbar and list box.
    # we are taking the root window as the master for the Frame.
ScrollFrame = tk.Frame(root, height = 200 , width = 200 , background = 'white')
ScrollFrame.pack(pady = 20)
    
    # IN above statement we have created a Frame and put it on the root window with some padding in y direction.
    
    # Now we will make Scrollbar and a Listbox and will put them into the 'ScrollFrame'
    
    # making a scrollbar
scroll = tk.Scrollbar(ScrollFrame)
scroll.pack(side = tk.RIGHT, fill=tk.Y)

    # So, Now we wlil make a Listbox widget also which would be connected to the scrollbar
List = tk.Listbox(ScrollFrame , yscrollcommand = scroll.set)
    # here we need to use the x or yscrollcommand to set the scrollbar
    
    # now we need to insert some lines inside the Listbox so we can have the line out of the widget
for line in range(30):
    
    List.insert(tk.END,"Number "+str(line))
    
    # now we need to pack the Listbox 
List.pack(side = tk.LEFT)

# NOw we need to configure the scrollbar in which we  need to connect the command option to 
# the Listbox xview or yview accroding to the requirment.

scroll.config(command = List.yview)

    
    
"""
    Result: Here first we created a Frame to hold the scrollbar and Listbox and isolate them with other widget,
            
            Now here you will found that the scrollbar and the Listbox are a single widget because they are adjecent to each other and 
            isolated to other widget which are not inside the ScrollFrame.
            
            This is the best way to create a scrollbar for any other widget which supports the scrolling.
        
"""  
#------------------------------------------------------------------------------------------------

root.mainloop()
````

#### Making a List-Scrollbar in Horizontal and Vertical direction 

* To create the Horizontal scrollbar we will use the `orent = tk.HORIZONTAL`
* yscrollcommand and yview are used to scrolling in y direction.
* xscrollcommand and xview are used to scrolling in x direction.

````python
## creating a scrollbar to scroll the Listbox in a single isolated Frame in box direction.

root = tk.Tk()

root.title("Scrollbar Tutorial")
root.geometry("300x300")
root.iconbitmap("./bitmap icons/home.ico")
#------------------------------------------------------------------------------------------------

    # Now first of all we need to make a Frame in which we will place the scrollbar and list box.
    # we are taking the root window as the master for the Frame.
ScrollFrame = tk.Frame(root, height = 200 , width = 200 , background = 'white')
ScrollFrame.pack(pady = 20)
    
    # IN above statement we have created a Frame and put it on the root window with some padding in y direction.
    
    # Now we need to make two Scrollbars one for the Horizontal and one for vertical direction.
    
    # making a scrollbar for vertical direction
Yscroll = tk.Scrollbar(ScrollFrame)
Yscroll.pack(side = tk.RIGHT, fill=tk.Y)
    # packed it to the right side of the Scrollbar ,because it will be used to scroll in Y-direction.
    
    # making a scrollbar for horizontal direction
    # here we need to use the orient option to make the scrollbar for x direction 
Xscroll = tk.Scrollbar(ScrollFrame , orient = tk.HORIZONTAL)
Xscroll.pack(side = tk.BOTTOM, fill=tk.X)
    # packed it to the bottom of the Scrollbar, because it will be used to scroll in X-direction.
    


    # So, Now we wlil make a Listbox widget also which would be connected to the scrollbar
    # Here also we need to connect the both Scrollbars acoording to their usage
    
    # Yscroll bar will be set in the yscrollcommand because it is for y-direction
    # Xscroll bar will be set in the xscrollcommand because it is for x-direction,
List = tk.Listbox(ScrollFrame , yscrollcommand = Yscroll.set , xscrollcommand = Xscroll.set)
    
    # now we need to insert some lines inside the Listbox so we can have the line out of the widget
for line in range(30):
    
    List.insert(tk.END,"Number for the counting at "+str(line))
    
    # now we need to pack the Listbox 
List.pack(side = tk.LEFT)

# Now we will configure the Xscroll and Yscroll according to their roll. 

# Yscroll will be configured in commmand to the yview of the List
Yscroll.config(command = List.yview)

# xscroll will be configured in commmand to the xview of the List
Xscroll.config(command = List.xview)

# Now our work has completed for this tutorial, You can also use the additional option in this for more features.
    
"""
    Result: Here first we created a Frame to hold the scrollbar and Listbox and isolate them with other widget,
            
            Now here you will found that the scrollbar and the Listbox are a single widget because they are adjecent to each other and 
            isolated to other widget which are not inside the ScrollFrame.
            
            This is the best way to create a scrollbar for any other widget which supports the scrolling.
        
"""  
#------------------------------------------------------------------------------------------------

root.mainloop()
````
