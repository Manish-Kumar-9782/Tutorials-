## Checkbutton widget
Also called the Checkbox

The purpose of a checkbutton widget to allow the user to read and select a two-way choice. The graphic above shows how checkbuttons lool in the off (0) and on (1) state in one implementation. The indicator is the part of the checkbutton that shows its state, and the label is the text that appears beside it.

* You will need to create a control varable, an instance of the IntVar class, so your program can query and set the state of the checkbutton.
* You can also use event bindings to react to user actions on the checkbutton.
* You can disable a checkbutton. Theis changes its appearance to "grayed out" and makes it unresponsive to the mouse.
* You can get rid of the checkbutton indicator and make the whole widget a"push-push" button that looks recessed when it is set, and looks raised when it is cleared,

To create the checkbutton in an existing parent window or frame parent:
```ruby
Cb = tk.Checkbutton(parent, options.....)
```

#### 1. Indicatoron

Normally a checkbutton dispalys as its indicator a box that shows whether the checkbutton is set or not. You can get this behavior by setting indicatoron=1. However. if you  set indicatoron=0, the indicator disappears, and the entire widget becomes a push-push button that looks _raised_ when it is cleared and _sunken_ when it is set. You may want to increase the borderwidth value to make it easier to see the state of such control.

**if indicatoron = 0**

````python
# creating a root winodow

root = tk.Tk()

root.title("Checkbutton Tutorial")
root.geometry("400x300")
root.iconbitmap("./bitmap icons/home.ico")
#-------------------------------------------------------------------------------------------------------------------

# now we will make a Checkbutton (checkbox) object and pack it to root window.

    # Current Case: indicatoron = 0
    
    # in this we will make the checkbutton as the push-push button by setting the indicatoronl as 0.
    
    # We will make: 
        # checkbox1 as the default checkbox
        # checkbox2 as the push-push button
    
    # checkbox2: in this we will add the activeforeground and the activebackground for make it clearly visiable
    
checkbox1 = tk.Checkbutton(root, text = 'Male'  , indicatoron = 1).pack(side = tk.TOP , anchor = tk.NW)
checkbox2 = tk.Checkbutton(root, text = "Female" , indicatoron = 0, activeforeground = 'white', activebackground = 'blue' )
checkbox2.pack(side = tk.TOP , anchor = tk.NW)
   

    # Result: Here checkbox1 is have the default indicatoron but the checkbox2 has the push-push button effect by changing 
            # the indicatoron, and also we added acitveforeground and acitvebackground in the checkbox2.

    
#-------------------------------------------------------------------------------------------------------------------

root.mainloop()
````

#### 2. offrelief
By default, checkbuttons use the tk.RAISED relief style when the button is off (cleared); use this option to specify a different relief style to be displayed when the button is off.

Note: This effect will be visible when we use the indicatoron = 0 and make it push-push button.

````python
# creating a root winodow

root = tk.Tk()

root.title("Checkbutton Tutorial")
root.geometry("200x200")
root.iconbitmap("./bitmap icons/home.ico")
#-------------------------------------------------------------------------------------------------------------------

# now we will make a Checkbutton (checkbox) object and pack it to root window.

    # Current Case: offrelief = tk.FLAT || tk.GROOVE 
    
    # in this we will change the value of the offrelief option values and will make it as tk.FLAT or tk.GROOVE or any other relief style.
    
    # We will make: 
        # checkbox1 as the default checkbox
        # checkbox2 ---> offrelief  = tk.FLAT 
    
    # checkbox2:  in this we will add the activeforeground and the activebackground for make it clearly visiable
                # indicatoron made as 0, because offrelief effect will not be visible on the default indicatoron setting.
                # 
    
checkbox1 = tk.Checkbutton(root, text = 'Male'  ).pack(pady = 20)
checkbox2 = tk.Checkbutton(root, text = "Female" , offrelief = tk.FLAT , activeforeground = 'white', activebackground = 'blue',
                              bg= 'white' , indicatoron = 0)
checkbox2.pack(pady = 10)
   

    # Result: Here checkbox1 is have the default indicatoron but the checkbox2 has the push-push button effect by changing 
            # the indicatoron, and also we added acitveforeground and acitvebackground in the checkbox2 so that we can see the pressed
            # effect of the checkbutton when it is pressed.
            # offrelief is made as the tk.FLAT, Now this checkbox2 will be flat in the normal (cleared) state.

    
#-------------------------------------------------------------------------------------------------------------------

root.mainloop()
````

#### 3. overrelief
Use this option to secify a relief style to be displayed when the mouse is over the checkbutton;

Note: we already have seen this option in the button widget but here also we will see that how it works here.

````python
# creating a root winodow

root = tk.Tk()

root.title("Checkbutton Tutorial")
root.geometry("200x200")
root.iconbitmap("./bitmap icons/home.ico")
#-------------------------------------------------------------------------------------------------------------------

# now we will make a Checkbutton (checkbox) object and pack it to root window.

    # Current Case: overrelief
    
    # in this we will test the effect of the overrelief option on the both checkboxs, we will apply it on the both of them.
    # We will make: 
        # checkbox1 as the default checkbox + the overrelief effect
        # checkbox2 ---> offrelief  = tk.FLAT 
                        # Now all the setting remaining as they were in the offrelief section but here also we will add 
                        # overrelief setting also. 
    
    # checkbox2:  in this we will add the activeforeground and the activebackground for make it clearly visiable
                # indicatoron made as 0, because offrelief effect will not be visible on the default indicatoron setting.
                
    
checkbox1 = tk.Checkbutton(root, text = 'Male'  , overrelief = tk.GROOVE).pack(pady = 20)
checkbox2 = tk.Checkbutton(root, text = "Female" , offrelief = tk.FLAT , activeforeground = 'white', activebackground = 'blue',
                              bg= 'white' , indicatoron = 0 , overrelief = tk.GROOVE)
checkbox2.pack(pady = 10)
   

    # Result: Here we applied the overrelief effect on the both checkboxs and it works on both of them with same effect.
            # we added acitveforeground and acitvebackground in the checkbox2 so that we can see the pressed
            # effect of the checkbutton when it is pressed.
            # offrelief is made as the tk.FLAT, Now this checkbox2 will be flat in the normal (cleared) state.

    
#-------------------------------------------------------------------------------------------------------------------

root.mainloop()
````

#### 4. offvalue and onvalue , command , variable
`offvalue:` Normally, a chaekbutton's asociated control variable will be set 0 when it is cleared (off). You can supply an alternate value for the off state by seetting _offvalue_ to that value.

`onvalue:` Normllly, a checkbutton's asociated control variable will be set to 1 when it is set (on). You can supply an alternate vlaue for the on state by setting _onvalue_ to that value.


````python
# creating a root winodow
root = tk.Tk()

root.title("Checkbutton Tutorial")
root.geometry("200x200")
root.iconbitmap("./bitmap icons/home.ico")
#-------------------------------------------------------------------------------------------------------------------

# now we will make a Checkbutton (checkbox) object and pack it to root window.

    # Current Case: onvalue and offvalue
    
    # in this we will test custom and default values by using the checkbox1 and checkbox2
    # We will make: 
        # checkbox1 as the default checkbox + having the default values of offvalue and onvalue.
        # checkbox2 --->  we will set the custom value of offvalue and onvalue for this checkbox.
        
                        # and rest off the setting will be restored.

# Since we are adding the custom values to the checkbox and to grab the state or the value of current state of the button
# we need a variable which is called the control variable.

State1 = tk.IntVar()
State2 = tk.StringVar()

# after defining the control variable we need to make them use when the button is pressed.
# For that we need to make a function which will be called from the command option.

# Since we need the use labels to show the value of the current state of the checkbutton 
# but for this we need to create them.

Text1 =  tk.StringVar()
Text2 =  tk.StringVar()

Text1.set("Male button: ")
Text2.set("Female button:")
label1 = tk.Label(root, textvariable = Text1)
label2 = tk.Label(root, textvariable = Text2)

def show_state():
    
    """ 
    * This function will display the value of the state for both of the checkbuttons.
    * We need to use the label widget to display the value of the current state of the checkbutton. 
    """
    
    # now here we need to set the values after pressing the buttons
    Text1.set(f"Male : {State1.get()}")
    Text2.set(f"Female: {State2.get()}")
            
        
     # now we need to add:
        # variable, command , offvalue and onvalue on the corresponding checkboxes.
        
checkbox1 = tk.Checkbutton(root, text = 'Male' , variable = State1 , command = show_state).pack(pady = 20)
checkbox2 = tk.Checkbutton(root, text = "Female" , offvalue='Female_OFF' , onvalue = 'Female_ON' , variable = State2,
                          command = show_state)
checkbox2.pack(pady = 10)
label1.pack()
label2.pack()
   

    # Result: Here we have finished the work of the modifying the and accessing the onvalue and the offvalue of the current state 
            # of the checkboxes. now we can see the current state values on the root window which are showing by using the labels.
#-------------------------------------------------------------------------------------------------------------------

root.mainloop()
````

**Explanation of extra object**

Here we used some other things like Text1, Text2, label1, label2, State1, State2.

`Label1 and Label2:` They are used to display the current state value on the root window.

`Text1 and Text2:` These are the control valriable which are used to update the current state vlalue of the checkbutton on the labels. These variable are passed in the textvariable option.

`State1 and State2:` These are also control variable but these variable are used to access the values of the current state of the checkbuttons. They are also used in the _show_state_  function to get the current state value of the checkbutton.
