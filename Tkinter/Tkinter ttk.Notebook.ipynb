## Tkinter ttk Notebook
Notebook a widget of ttk which is the extension pack of the tkinter. Notebook provide us the ability to work on a common area with different pages or tabes, this tabes contains different content and they can be display on the common area at a time.

This Notebook widget is very useful when we want multiple content on the same area with just a switch of button. 

* Each tab contain a child which is a Frame. This frame can contain anything inside it.
* To add a frame into the Notebook we need to use .add() method on a Notebook object.
* Each tab has its own set of options that control its appearane and behavior.
* Each tab has its own unique id which is calle `tabId`.
* if we pass 'current' string to select a tab in `Notebook.select([tabId])`, then we will select the currently displayed tab.
* To see the current number of tabs displayed, we can use `Notebook.index()` method.

Now we will make a Notebook but before that we need to import some module

```python
# To make a Notebook we need to import the ttk.Notebook along with tkinter
import tkinter as tk
from tkinter import ttk
```

```python
# Now first of all we need to make a root window.

root= tk.Tk()
root.geometry("400x400")
root.title('Notebook Tutorial')

# Now we need to create a Notebook object, and it will be placed in the root window.
notebook = ttk.Notebook(root)
notebook.pack()

# Now after packing the notebook we need to make some Frame which will be used as the tabs of the Notebook
tracks = tk.Frame(root, bg = 'red' , height = 400, width = 400)
keeps = tk.Frame(root, bg = 'green', height = 400, width = 400)
plans = tk.Frame(root, bg = 'blue', height = 400, width = 400)

# Now we need to add these Frame in the Notebook.
# Note: we should note here that we are note packing any frame here. they will be packed by the notebook itself.
notebook.add(tracks, text = 'Tracks')
notebook.add(keeps, text = 'Keeps')
notebook.add(plans, text = 'Plans')


root.mainloop()
```

Now we have created a simple programme which contains the Notebook. In this notebook we have some tabs and these tabs contains Frame as the child of the tabs. Now we can put anything in those Frame as we want. 

```python
# Now first of all we need to make a root window.

root= tk.Tk()
root.geometry("400x400")
root.title('Notebook Tutorial')

# Now we need to create a Notebook object, and it will be placed in the root window.
notebook = ttk.Notebook(root)
notebook.pack()

# Now after packing the notebook we need to make some Frame which will be used as the tabs of the Notebook
tracks = tk.Frame(root, bg = 'red' , height = 400, width = 400)
tracks.pack(side = tk.TOP)

keeps = tk.Frame(root, bg = 'green', height = 400, width = 400)
keeps.pack(side = tk.TOP)

plans = tk.Frame(root, bg = 'blue', height = 400, width = 400)
plans.pack(side = tk.TOP)

# Now we need to add these Frame in the Notebook.
# Note: we should note here that we are note packing any frame here. they will be packed by the notebook itself.
notebook.add(tracks, text = 'Tracks',  sticky = 'nw')
notebook.add(keeps, text = 'Keeps', sticky = 'w')
notebook.add(plans, text = 'Plans', sticky = 'sw')


root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

root = tk.Tk()

style = ttk.Style()
print("widget layout: ",style.layout("TNotebook"))
print("widget layout: ",style.element_options("Notebook.client"))
style.configure('lefttab.TNotebook', tabposition='wn' , background = '#2fec3f', tabmargins = (0))
#style.configure('TNotebook.Tab', background ='black')

#print(dir(style))



notebook = ttk.Notebook(root, style='lefttab.TNotebook')


# if we want to add a some gap b/w two tabs then we can use a seperator
sep = ttk.Separator(notebook, orient = 'vertical')

print(notebook['style'])
print(notebook.tabs())
#print(notebook.tab(0, option='background'))
f1 = tk.Frame(notebook, bg='red', width=200, height=200)
f2 = tk.Frame(notebook, bg='blue', width=200, height=200)

notebook.add(f1, text='Frame 1')
# Now add the separator b/w the two tabs/Frame
notebook.add(sep)
notebook.add(f2, text='Frame 2')

notebook.grid(row=0, column=0, sticky="nw")



print(notebook.tabs())

print(notebook.tab('.!notebook.!frame'))
root.mainloop()
```

```python
import tkinter as tk
import tkinter.ttk as ttk

def stylename_elements_options(stylename):
    '''Function to expose the options of every element associated to a widget
       stylename.'''
    try:
        # Get widget elements
        style = ttk.Style()
        layout = str(style.layout(stylename))
        print('Stylename = {}'.format(stylename))
        print('Layout    = {}'.format(layout))
        elements=[]
        for n, x in enumerate(layout):
            if x=='(':
                element=""
                for y in layout[n+2:]:
                    if y != ',':
                        element=element+str(y)
                    else:
                        elements.append(element[:-1])
                        break
        print('\nElement(s) = {}\n'.format(elements))

        # Get options of widget elements
        for element in elements:
            print('{0:30} options: {1}'.format(
                element, style.element_options(element)))

    except tk.TclError:
        print('_tkinter.TclError: "{0}" in function'
              'widget_elements_options({0}) is not a regonised stylename.'
              .format(stylename))

stylename_elements_options('TNotebook')
```

## Get ttk widget's element options

```python
import tkinter as tk
import tkinter.ttk as ttk

def iter_layout(layout, tab_amnt=0, elements=[]):
    """Recursively prints the layout children."""
    el_tabs = '  '*tab_amnt
    val_tabs = '  '*(tab_amnt + 1)

    for element, child in layout:
        elements.append(element)
        print(el_tabs+ '\'{}\': {}'.format(element, '{'))
        for key, value in child.items():
            if type(value) == str:
                print(val_tabs + '\'{}\' : \'{}\','.format(key, value))
            else:
                print(val_tabs + '\'{}\' : [('.format(key))
                iter_layout(value, tab_amnt=tab_amnt+3)
                print(val_tabs + ')]')

        print(el_tabs + '{}{}'.format('} // ', element))

    return elements

def stylename_elements_options(stylename, widget):
    """Function to expose the options of every element associated to a widget stylename."""

    try:
        # Get widget elements
        style = ttk.Style()
        layout = style.layout(stylename)
        config = widget.configure()

        print('{:*^50}\n'.format(f'Style = {stylename}'))

        print('{:*^50}'.format('Config'))
        for key, value in config.items():
            print('{:<15}{:^10}{}'.format(key, '=>', value))

        print('\n{:*^50}'.format('Layout'))
        elements = iter_layout(layout)

        # Get options of widget elements
        print('\n{:*^50}'.format('element options'))
        for element in elements:
            print('{0:30} options: {1}'.format(
                element, style.element_options(element)))

    except tk.TclError:
        print('_tkinter.TclError: "{0}" in function'
                'widget_elements_options({0}) is not a regonised stylename.'
                .format(stylename))

widget = ttk.Button(None)
class_ = widget.winfo_class()
stylename_elements_options(class_, widget)
```
