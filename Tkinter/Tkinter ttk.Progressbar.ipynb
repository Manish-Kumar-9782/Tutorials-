# ProgressBar
In tkinter module Progressbar is extension of the tkinter widget it comes in ttk class, the purpose of this widget is to reassure the user that something is happening.\
It can be operte in one two modes.
1. determinate mode: the widget shows an indicator that moves from beginning to end under program control.
2. Indeterminate mode: the widget is animated so the user will believe that something is in progress. In this mode the indicator bounces back and forth b/w the ends of the widget.

Options: 
* class_
* cursor
* length 
* maximum
* mode
* orient
* style
* takefocus
* variable

see the documentation for detailed description....

To make a progress bar first of all we need to make a root window and then we need to put progressbar in that window. first we will se the determinate mode.

### 1. Determinate mode

```python
import tkinter as tk
from tkinter import ttk
import time
```

```python
root = tk.Tk()
root.geometry('600x400')
root.title("Progress Bar")

# now here we need to to make a progressbar object.
# the master window for this progressbar will be root window.
progress = ttk.Progressbar(root, length = 300)

# before packing the progress bar we will make a label for progress bar.
label = tk.Label(root, text = 'Progress Bar: ')
# this will be packed before the progress bar


# Now we will add another label to display the progress value.
progress_text = tk.Label(root, text = '')


label.pack(pady = 20, side = 'left',  anchor = 'n')
progress.pack(pady = 20, side = 'left',  anchor = 'n')
progress_text.pack(pady = 20, side = 'left',  anchor = 'n')
# Now both progress bar and the label placed together 

# now we need to run the progress bar, to run the progress bar we need to set some values as the time passes.
# for this demonstration we will add 100/60 percent for one second.

root.update_idletasks()
def progress_value():
    
    progress['value'] += 100/60
    a = progress['value']
    
    # here we need to update the progress_value text as it passes.
    progress_text.config(text = f"{a:.2f} Completed")
    
    progress.after(1000, progress_value)
    
progress_value()


root.mainloop()
```

Now we can make a progress bar with label, we can put these three thing into a single frame and then use this frame any where in any application. 

```python
def ProgressBar(master , prefix="Progress", suffix = 'Completed', increment = None):
    
    """
    master: In this Progressbar will be displayed.
    prfix: This will be put before the Progress bar.
    suffix: This will be put after the progress bar, and in general it is used to show that how much progress has been completed.
    """
    
    Progress_after = None
    
    # Now first we will import the necessary things for the ProgressBar
    from tkinter import Label, Frame
    from tkinter import ttk
    
    # we need a Frame to put all the things on that.
    # Label are used to put and display the prefix and suffix.
    # Progresbar to create the progress.
    
    # first of all we need to make a Frame
    
    Progress_Frame = Frame(master, height = 60, width= 300)
    Progress_Frame.pack()
    s = ttk.Style()
    s.theme_use('clam')
    s.configure("red.Horizontal.TProgressbar", foreground='red', background='red')
    # Now we need to put all the thing in this progress_bar frame
    Progress_Bar = ttk.Progressbar(Progress_Frame, length  = 200 , style = 'red.Horizontal.TProgressbar')
    Prefix_label = Label(Progress_Frame, text = prefix)
    Suffix_label = Label(Progress_Frame, text = suffix)
    
    # Now we need to pack all thing in order, Prefix_label, Progress_Bar, Suffix_label
    Prefix_label.pack(side = 'left')
    Progress_Bar.pack(side = 'left')
    Suffix_label.pack(side = 'left')
    
    # Now packing is done now we need to run the Progress bar for that we need to use a w.after() method.
    
    def update_progress():
        global Progress_after
        # Now first off all we need add the increment in each call
        Progress_Bar['value'] += increment
        
        # Now after adding the increment our progress bar updated, now we need to show the incremented value on the label.
        # this will be updated on the Suffix label
        Suffix_label.config(text = f"{Progress_Bar['value']:.2f} " + suffix)
        
        # now after updating both thing we need to call this function in .after method
        
        #print('\r',Progress_Bar['value'] , end = '\r')
        
        if Progress_Bar['value'] >= 100:
            # Now when our progress reaches to 100% then we need to stop this .after method using a global method.
            print("progress completed")
            Suffix_label.after_cancel(Progress_after)
        else:
            # for now trial we will call this method in every 1 sec. 
            Progress_after = Suffix_label.after(1000, update_progress)
            # change the sleep time according to your need/ project
        
    update_progress()
    
```

##### Running the progress bar with an root window.

```python
root = tk.Tk()
root.geometry('400x400')

PFrame = tk.Frame(root)
PFrame.pack(pady = 20)

ProgressBar(PFrame, increment=100/10)


root.mainloop()
```

### 2. Indeterminate Progress bar
Indeterminate Progress bar is different from the determinate progress bar, in this mode we will just see a single dot not a complete full length bar for progress. This will single dot will be animated through out the progress path. 

To make a indeterminate progress we just need to change a the mode of the progress bar, `mode=indeterminate`

```python
def ProgressBar(master , prefix="Progress", suffix = 'Completed', increment = None):
    
    """
    master: In this Progressbar will be displayed.
    prfix: This will be put before the Progress bar.
    suffix: This will be put after the progress bar, and in general it is used to show that how much progress has been completed.
    """
    
    Progress_after = None
    
    # Now first we will import the necessary things for the ProgressBar
    from tkinter import Label, Frame
    from tkinter import ttk
    
    # we need a Frame to put all the things on that.
    # Label are used to put and display the prefix and suffix.
    # Progresbar to create the progress.
    
    # first of all we need to make a Frame
    
    Progress_Frame = Frame(master, height = 60, width= 300)
    Progress_Frame.pack()
    
    
    # Now we need to put all the thing in this progress_bar frame
    Progress_Bar = ttk.Progressbar(Progress_Frame, length  = 200 , mode= "indeterminate")
    Prefix_label = Label(Progress_Frame, text = prefix)
    Suffix_label = Label(Progress_Frame, text = suffix)
    
    # Now we need to pack all thing in order, Prefix_label, Progress_Bar, Suffix_label
    Prefix_label.pack(side = 'left')
    Progress_Bar.pack(side = 'left')
    Suffix_label.pack(side = 'left')
    
    # Now packing is done now we need to run the Progress bar for that we need to use a w.after() method.
    
    def update_progress():
        global Progress_after
        # Now first off all we need add the increment in each call
        Progress_Bar['value'] += increment
        
        # Now after adding the increment our progress bar updated, now we need to show the incremented value on the label.
        # this will be updated on the Suffix label
        Suffix_label.config(text = f"{Progress_Bar['value']:.2f} " + suffix)
        
        # now after updating both thing we need to call this function in .after method
        
        #print('\r',Progress_Bar['value'] , end = '\r')
        
        if Progress_Bar['value'] >= 100:
            # Now when our progress reaches to 100% then we need to stop this .after method using a global method.
            print("progress completed")
            Suffix_label.after_cancel(Progress_after)
        else:
            # for now trial we will call this method in every 1 sec. 
            Progress_after = Suffix_label.after(1000, update_progress)
            # change the sleep time according to your need/ project
        
    update_progress()
    
```

```python
root = tk.Tk()
root.geometry('400x400')

PFrame = tk.Frame(root)
PFrame.pack(pady = 20)

ProgressBar(PFrame, increment=100/10)


root.mainloop()
```
