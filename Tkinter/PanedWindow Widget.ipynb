## The PanedWindow Widget
The purpose of the _PanedWindow_ widget is to five the application's user some control over how space is divided up within the application.

A _PanedWindow_ is somewhoat like _Frame_: it is a container for _child widgets_. Ech **PanedWindow** a widget cotains a horizontal or vertical stack of child widgets. Using the mouse, the user can drag the boundaries b/w the child widgets back and forth.

* `handle:`  You may choose tho display handle within the widget. A handle is a small square that the user can drag with mouse.
* `sash:`  You may choose to make _sashes visible_. A sash is a bar placed b/w the child widgets.
* `pane:` A _pane_ is the area occupied by one child widget.

To crate a new **PanedWindow** widget as the child o fa root window or frame named parent:
```ruby
PanedWindow = tk.PanedWindow(root, options...)
```

`Common Options:`
* bg or background
* bd or borderwidth
* cursor
* height and width
* relief

`Additional Options:`
* handlepad
* handlesize
* opaqueresize
* orient
* sashpad
* sashrelief
* sashhandle
* showhandle

#### Using: handlepad and handlesize of PanedWindow

`handlepad:` Use this option shpecify the distance b/w the handle and the end of the sash. 
* for _orient =  tk.VERTICAL_. this is the distance b/w the left end of the sash and the handle; 
* for  _orient = tk.HORIZONTAL_,  it is the distance b/w the top of the sash and the handle. The default value is eight pixles;

`handlesize:` Use this option specify the size of the handle, which is always a square;

````python
# creating a root winodow
import tkinter as tk

root = tk.Tk()

root.title("Menu Tutorial")
root.geometry("400x300")
root.iconbitmap("./bitmap icons/home.ico")
#-------------------------------------------------------------------------------------------------------------------

    # Here, we will make PanedWindow widget
    
# Now we will make the PanedWindow.

# NOte: You should use the background and borderwidth option to disinguish the PanedWindow from others.
# color1: #6edba3
# color2: #d5e36f
# color3: #eba07a


# This is over first PanedWindow, we will place this left of the root window
    
    #Addition:
        # Now in this we will add a sashpad (can be called sash padding)
        # Adding handlepad and showhandle

Paned1 = tk.PanedWindow(root, background = '#6edba3' , borderwidth = 5 , 
                        relief = 'raised' , orient = tk.HORIZONTAL , width = 50,
                         handlepad = 100, showhandle = True)
Paned1.pack(fill = tk.BOTH , expand=1)

# to make Paned visible we need to put some Label or other widget
# Making a Label for the first PanedWindow, So the master for the Label will be Paned1
label1 = tk.Label(Paned1, text = 'Label1 IN Paned1') 
# Now we need to pack this , but we need to pack this inside the Paned1 so, for that we need to use the .add() method of a PanedWindow
Paned1.add(label1)


# NOw we will put some other Label inside the Second PanedWindow

# This is over second PanedWindow, we will place this right of the root window

     #Addition:
            # Now we will add showhandle as True;
            #
        
Paned2 = tk.PanedWindow(root, background = '#d5e36f' , borderwidth = 5 , 
                        relief = 'sunken', orient = tk.VERTICAL, width = 50,
                       showhandle = True  , handlesize = 40)
Paned1.add(Paned2)

label2 = tk.Label(Paned2, text = 'Label2 IN Paned2')
label3 = tk.Label(Paned2, text = 'Label3 IN Paned2')

Paned2.add(label2)
Paned2.add(label3)

"""
Result: Now in this we have added two new things one is handlepad and the other is handlesize.
        To make this two effect visible, first we need to set the showhandle as True;
        
        handlesize: this will set the size of the handle, shape of the handle is square
        handlepad: this will define the position of the handle on the sash line. the postion of the x and y point of the handle 
                    will depend on the orient of the PanedWindow.
                    
        In Paned1 we have used the handlepad.
        In Paned2 we have used the handlesize.
                    
"""

#-------------------------------------------------------------------------------------------------------------------

root.mainloop()
````

#### .forget(child)  & .paneconfig()
This function removes a child widget from the container.

````python
# creating a root winodow
import tkinter as tk

root = tk.Tk()

root.title("Menu Tutorial")
root.geometry("400x300")
root.iconbitmap("./bitmap icons/home.ico")
#-------------------------------------------------------------------------------------------------------------------

    # Here, we will make PanedWindow widget
    


Paned1 = tk.PanedWindow(root, background = 'red' , borderwidth = 5 , 
                        relief = 'raised' , orient = tk.VERTICAL , width = 50,
                          sashrelief = 'raised' , sashwidth = 10, opaqueresize = False)
Paned1.pack(fill = tk.BOTH , expand=1)

label1 = tk.Label(Paned1, text = 'Label1 IN Paned1') 
Paned1.add(label1)
        
Paned2 = tk.PanedWindow(root, background = '#d5e36f' , borderwidth = 5 , 
                        relief = 'flat', orient = tk.VERTICAL, width = 50,
                       sashwidth = 50 )
Paned1.add(Paned2)

label2 = tk.Label(Paned2, text = 'Label2 IN Paned2')
label3 = tk.Label(Paned2, text = 'Label3 IN Paned2')

Paned2.add(label2)
Paned2.add(label3)

#------------------------------------------------FORGETTING A WIDGET------------------------------------------------

# Now first we have to create a function which will use be used to call the forget method of the Panedwindow
def forget_child():
    # Here we will forget the first label of Paned2
    Paned2.forget(label2)

# Now we have created a function which will forget the label2
# Now we need to configure the forget_button for the commoand option. for that we will create a button.

# Here we will make a button an we will place it inside the Paned1 but it will forget Paned2 widget
# Making a button which will run the forget_child method.

forget_button = tk.Button(Paned1 , text="forget" , activebackground = "#e24ff0", activeforeground = 'white' , command = forget_child)
Paned1.add(forget_button , before = label1)



Paned1.paneconfig(forget_button , minsize= 30)

# Here:
        # minsize: this will set the minimum hight or width according to the orientation of the PanedWindow.
    
#-------------------------------------------------------------------------------------------------------------------

"""
Result: Now As we use the forget from the application (in Paned1 section) we can see the the label2 (Label2 IN Paned2) is not visible
        This is because when we use the forget button it forget this Label2 and removes from the continer.
        
"""

#-------------------------------------------------------------------------------------------------------------------

root.mainloop()
````

#### .remove(child) & .paneconfig(child, option...)
remove method is similar to the forget function.

````python
# creating a root winodow
import tkinter as tk

root = tk.Tk()

root.title("Menu Tutorial")
root.geometry("400x300")
root.iconbitmap("./bitmap icons/home.ico")
#-------------------------------------------------------------------------------------------------------------------

    # Here, we will make PanedWindow widget
    


Paned1 = tk.PanedWindow(root, background = 'red' , borderwidth = 5 , 
                        relief = 'raised' , orient = tk.VERTICAL , width = 50,
                          sashrelief = 'raised' , sashwidth = 10, opaqueresize = False)
Paned1.pack(fill = tk.BOTH , expand=1)

label1 = tk.Label(Paned1, text = 'Label1 IN Paned1') 
Paned1.add(label1)
        
Paned2 = tk.PanedWindow(root, background = '#d5e36f' , borderwidth = 5 , 
                        relief = 'flat', orient = tk.VERTICAL, width = 50,
                       sashwidth = 50 )
Paned1.add(Paned2)

label2 = tk.Label(Paned2, text = 'Label2 IN Paned2')
label3 = tk.Label(Paned2, text = 'Label3 IN Paned2')

Paned2.add(label2)
Paned2.add(label3)

#------------------------------------------------FORGETTING A WIDGET------------------------------------------------

# Now in this section we will create two function one for the forget() method and another for the remove() function.

def forget_child():
    # Here we will forget the first label of Paned2
    Paned2.forget(label2)


def remove_child():
    
    # here we will remove the second label of Paned2
    Paned2.remove(label3)
    
# Now we have created a function which will forget the label2
# Now we need to configure the forget_button for the commoand option. for that we will create a button.


# creating forget and remove button

forget_button = tk.Button(Paned1 , text="forget" , activebackground = "#e24ff0", activeforeground = 'white' , command = forget_child)
Paned1.add(forget_button , before = label1)

remove_button = tk.Button(Paned1, text ='remove' , activebackground = '#e24ff0', activeforeground = 'white' , command = remove_child)
Paned1.add(remove_button , after = forget_button)



Paned1.paneconfig(forget_button , minsize= 30 , sticky = tk.NW)
Paned1.paneconfig(remove_button, minsize = 30 , after = forget_button , sticky = tk.NW)

# Here:
        # minsize: this will set the minimum hight or width according to the orientation of the PanedWindow.
    
#-------------------------------------------------------------------------------------------------------------------

"""
Result: Now As we use the forget from the application (in Paned1 section) we can see the the label2 (Label2 IN Paned2) is not visible
        This is because when we use the forget button it forget this Label2 and removes from the continer.
        
"""

#-------------------------------------------------------------------------------------------------------------------

root.mainloop()
````

#### panecget(child, option)

This method retrieves the value of a achild widget configuration option, where _child_ is the child widget and _option_ is the name of the option as a string.

````python
# creating a root winodow
import tkinter as tk

root = tk.Tk()

root.title("Menu Tutorial")
root.geometry("400x300")
root.iconbitmap("./bitmap icons/home.ico")
#-------------------------------------------------------------------------------------------------------------------

    # Here, we will make PanedWindow widget
    


Paned1 = tk.PanedWindow(root, background = 'red' , borderwidth = 5 , 
                        relief = 'raised' , orient = tk.VERTICAL , width = 50,
                          sashrelief = 'raised' , sashwidth = 10, opaqueresize = False)
Paned1.pack(fill = tk.BOTH , expand=1)

label1 = tk.Label(Paned1, text = 'Label1 IN Paned1') 
Paned1.add(label1)
        
Paned2 = tk.PanedWindow(root, background = '#d5e36f' , borderwidth = 5 , 
                        relief = 'flat', orient = tk.VERTICAL, width = 50,
                       sashwidth = 50 )
Paned1.add(Paned2)

label2 = tk.Label(Paned2, text = 'Label2 IN Paned2')
label3 = tk.Label(Paned2, text = 'Label3 IN Paned2')

Paned2.add(label2)
Paned2.add(label3)

#-----------------------------------------------FRAME CONTAINER FO THE BUTTONS--------------------------------------

# Here the master will be Paned1, because we need this Frame inside the Paned1.
Frame1 = tk.Frame(Paned1 , background = 'white', padx=10, pady = 10 , height = 30 , width = 300)
Paned1.add(Frame1 , before = label1)


# NOte: now we will put all the button inside this Frame1. OK.....

#------------------------------------------------FORGETTING A WIDGET------------------------------------------------

# Now in this section we will create two function one for the forget() method and another for the remove() function.

# method to forget a child
def forget_child():
    # Here we will forget the first label of Paned2
    Paned2.forget(label2)

# method to remove a child
def remove_child():
    
    # here we will remove the second label of Paned2
    Paned2.remove(label3)

# methd to get the Panes list of a PanedWindow
def get_panes():
    
    print("Paned1 Panes Details:\n",Paned1.panes())
    print("\nPaned2 Panes Details:\n",Paned2.panes())
    

# method to get a configuration details of a panes child widget.

def get_config():
    
    # In this we will get the Frame1 details
    
    config_list = ["after", "before","height" , "minsize" , "padx" , "pady" , "sticky" , "width"]
    
    Paned2.forget(label2)
    Paned2.forget(label3)
    
    var = tk.StringVar() # in this we will store the config details and pass to the Message widget.
    config_msg = tk.Message(Paned2 , textvariable = var)

    msg = f"Panes Config list of Frame1:\n{'_'*30}\n"
    
    for option in config_list:
        
        msg = msg + f"{option}:    {Paned1.panecget(Frame1,option)} \n" 
    
    var.set(msg)
    # Now we need to add the Message inside the Paned2
    
    Paned2.add(config_msg)
    
    
# Now we have created a function which will forget the label2
# Now we need to configure the forget_button for the commoand option. for that we will create a button.


# creating forget and remove button

forget_button = tk.Button(Frame1 , text="forget" , activebackground = "#e24ff0", activeforeground = 'white' , command = forget_child)
forget_button.pack(side = 'left' , anchor = 'nw' , padx = 10)

remove_button = tk.Button(Frame1, text ='remove' , activebackground = '#e24ff0', activeforeground = 'white' , command = remove_child)
remove_button.pack(side = 'left' , anchor = 'nw' , padx = 10)

# Here in above code we have used the .paneconfig() method but now we dont need that 
# Since now these buttons are not inside the PanedWindow, Now they are now inside the Frame1 

# Paned1.paneconfig(forget_button , minsize= 30 , sticky = tk.NW)
# Paned1.paneconfig(remove_button, minsize = 30 , after = forget_button , sticky = tk.NW)

# Now we will create Panes button to call the panes() method.

Panes_button = tk.Button(Frame1, text ='Panes' , activebackground = '#e24ff0', activeforeground = 'white' , command = get_panes)
Panes_button.pack(side = 'left' , anchor = 'nw' , padx = 10)

# Now we will add the panecget button inside the Frame1

get_config_button =  tk.Button(Frame1, text ='get config' , activebackground = '#e24ff0', activeforeground = 'white' , 
                               command = get_config)
get_config_button.pack(side = 'left' , anchor = 'nw' , padx = 10)
#-------------------------------------------------------------------------------------------------------------------

"""
Result: Now As we use the forget from the application (in Paned1 section) we can see the the label2 (Label2 IN Paned2) is not visible
        This is because when we use the forget button it forget this Label2 and removes from the continer.
        
        Panes_button: This button (Panes) will print the list of Paned1 and Paned2 child widgets. The list will be displayed on the 
                      Console.
        
"""

#-------------------------------------------------------------------------------------------------------------------

root.mainloop()
````

#### identify(x,y)
For a given location (x,y) in window coordinates, this method returns a value that describe the feature at that location.

* if the feture is a child window, the method returs an empty stirinf.
* if the feature is a sash, the method returns a tuple _(n, 'sash')_ where n is 0 for the first sash, 1 for the second  and so on...
* if the feature is a handle , the method returns a tuple _(n,'handle')_ where n is 0 for the firt handle, 1 for the second and so on.
