```python
import tkinter as tk
```

```python
root = tk.Tk()
 # this section is the most fundamental part in which we can add some more widgets and give their property.
    # In this we add a Label to add a label we need one things which is master window 
    # This master window is the parent window in which the label is going to placed. 
    # like every widget a Label can have many option which can be usefull to make it stylish.
    
label = tk.Label(root,text="hello world!")

# In this label function, the very first argument is for the master window, here the master window is root in which this 
# label is going to display. there are other option also which can be seen from the below link

label.pack()

# here we used one additional function on the label is pack() which is used to pack the widget into the master window
# if we dont pack the widget (label) then nothing will be appear on the master window

root.mainloop()
```

 <p style ="font-family:serif; font-size:17px; "> Tkinter has three built-in layout managers: the pack, grid, and place managers. The grid geometry manager places widgets in a two dimensional grid. </p>

### 1.Label options and attributes



|Options|Discriptions |
|:------------------|:-------|
|**activebackgriound**|Background coloe to be displayed when the mouse is ouver the widget. |
|**actibeforegriound** |Foreground color to be displayed when the mouse is ouver the widget. |
|**anchor**|This options controls where the text is positoned, if the the widget has more space than the text needs. The default is *anchor=tk.CENTER*, which centers the text in the available space. |
|**bg** or **background** |The background color of the label area.|
|**bitmap**|Set this option equal to a bitmap or image objext and the label will display that graphic.|
|**bd** or **borderwidth**|Width of the border around the label. The default value is 2px.|
|**compound**|if uou would like the **Label** widget to display both text and a graphic, the *compound* iption specifies the relative orientation of the graphic relative to the text. Values may be any of **tk.LEFT, tk.RIGHT, tk.CENTER, tk.BOTTOM, or tk.TOP**. for example uf you specify **compound=BOTTOM,** the graphic will be displayed below the text. |
|**cursor**|Cursor that appears when the mouse is over the label.|
|**disabledforeground**|The foreground color to be displayed when the widget's **state** is **tk.DISABLED.**|
|**font**|if you are displaying thext in this label (with *text* or *textvariable*, the **font** option specifies in what font that text will ve diaplayed|
|**fg** or **foreground** |if you are displaying text or a bitmap in this label, this option specifies the color of the text. if you are displaying bitmap, this is the color that will appear at the position of the 1-bits in the bitmap.|
|**height**|Height of the label in *lines* (not pixles). if this is not set, the label will be sozed to fit its contents.|
|**highlightbackground**|Color of the focus highlight when the widget does not have focus.|
|**highlightcolor**|The color of the focus highlight when the widget has focus.|
|**highlightthickness**|Thickness of the highlight.|
|**image**|To display a static image in the label widget, set this option to an image object.|
|**justify**|Specifies how multiple lines of text will be aligned with respec to each other: **tk.LEFT** for flush left, **tk.CENTER** for centered, **tk.RIGHT** for right-justified. |
|**padx and pady**|**padx:** Extra space added to the left and right of the text within the widget. **pady:** Extra space added above and below the text within the widget. by default both values are 1.|
|**relief**|Specifies the appearance of a decorative borader around the label. The default is **tk.FLAT**; check other relief in that perticular page|
|**state**|By defult, an **Entry** widget is in the **tk.NORMAL** state. set **tk.DISABLED** to make it unresponsive to mouse events. The state will be **tk.ACTIVE** when the mouse is over the widget.|
|**takefocus**|NOrmally, focus does not cycke through **Label** widgets; "Focus routing keyboard input". if you want this widget to be visited by the focus, set **takefocus=1** |
|**text**|To display one or more lines of thext in a label widget, set this option a string containing the text. internal newlines ('\n') will fouce a line break.|
|**textvariable**|To slave the text displayed in a label widget to a control variable of class **StringVar** set this option to that variable. "Control variables: the values behind the widgets"|
|**underline**|You can display an underline ( _ ) below the nth letter of the text, counting from 0, by setting this option to n. The defaults is **underlines=-1** which means no underlining.  |
|**width**|Width of the label in *characters* (not pixles). if this option is not set, the label will be sized to fit its contents.|
|**wraplength**|You can limit the number of characters in each line by setting this option to the desired number. The defauylt value, 0, means that lines will be broken only at newlines.|

### 1.anchor

<p style="font-family:serif; font-size:20px;"> The anchor attribute is used to control display of the text positon inside the Label. it has the absolute position value which defines the absolute position for the given text. <br><br>
    The anchor values are given below.
    <ul>
        <li> tk.NW </li>
        <li> tk.N </li>
        <li> tk.NE </li>
        <li> tk.W </li>
        <li> tk.E </li>
        <li> tk.SW </li>
        <li> tk.SE </li>
        <li> tk.S </li>
        <li> tk.CENTER </li>
    </ul>     
</p>

we will see each of them individually

```python
# first of all we need to make the blank window we needo to import the tkinter as tk which is imported at the very first
# cell of this file

# making the root window which will be the parent window for the Label widget.
root = tk.Tk()
# before adding some Label we need to make some property of the main parent window.
# we need to add some properties like
# titel of the window, geometry for now we will add these two property.
root.geometry("710x520")
# root.geometry specify the initial size of the root window.
root.title("Label anchor Tutorial")
# root.title adds the title for the root window which is displayed at the title bar

# now we will add the anchors
# each anchor is placed with thir full name of the position.

label_config = {'bg':"#cafcf6",'fg':'Blue', 'height':10, 'width':30, 'borderwidth':5, 'relief':tk.RIDGE }
#label_config is the configuration for the label which share the common attributes.

#=========================================================================================================
#North west 
north_west_label = tk.Label(root, text="North west", anchor=tk.NW, cnf=label_config)
north_west_label.grid(row=0, column=0,padx=5,pady=5)

#North 
north_label = tk.Label(root, text="North", anchor=tk.N, cnf=label_config)
north_label.grid(row=0, column=1,padx=5,pady=5)

#North east
north_east_label = tk.Label(root, text="north east", anchor=tk.NE, cnf=label_config)
north_east_label.grid(row=0, column=2,padx=5,pady=5)

#West
west_label = tk.Label(root, text="west", anchor=tk.W, cnf=label_config)
west_label.grid(row=1, column=0,padx=5,pady=5)

#East
east_label = tk.Label(root, text="east", anchor=tk.E, cnf=label_config)
east_label.grid(row=1, column=2,padx=5,pady=5)

#South west
south_west_label = tk.Label(root, text="south west", anchor=tk.SW, cnf=label_config)
south_west_label.grid(row=2, column=0,padx=5,pady=5)

#South
south_label = tk.Label(root, text="south", anchor=tk.S, cnf=label_config)
south_label.grid(row=2, column=1,padx=5,pady=5)

#South east
south_east_label = tk.Label(root, text="south east", anchor=tk.SE, cnf=label_config)
south_east_label.grid(row=2, column=2,padx=5,pady=5)

#Center
center_label = tk.Label(root, text="center", anchor=tk.CENTER, cnf=label_config)
center_label.grid(row=1, column=1,padx=5,pady=5)

#==========================================================================================================

# Note: we have used the grid layout manager to place the label widgets. we will se the layout manager in a 
# seperate tutorial page 
root.mainloop()
```

### 2. justify, padx, pady, text, textvariable

<p style="font-family:serif; font-size:20px;"> 
    <ul style="font-family:serif; font-size:20px;">
        <li>Jusity: justify is used to make the multiple line of text in an align format </li>
        <li>padx: padx adds the additional space in to the left and right both side </li>
        <li>pady: pady adds the additional space to the top and bttom of the text.</li>
        <li>text: text is the parameters which holds the string which is to be displayed on the label.</li>
        <li>textvariable: textvariable is a control variable parameters which holds the string to use this parameters we need to use the StringVar and we would set some string in it.</li>
    </ul>
</p>

<span style="font-family:serif; font-size:20px; color:red; ">Note:</span> 
<span style="font-family:serif; font-size:20px;"> I am not defining those line of code which have been seen already in above code if you are not familier with the line then see it on the above example.
    <br> <br>In this we will create four label in which they are Normallabel, justifylabel, padxlabel, padylabel.
</span>

```python
# as we have seen we need to make the a new window for the label again

root = tk.Tk()
root.geometry("800x400")
root.title("justify, padx and pady")

label_config = {'bg':"#cafcf6",'fg':'Blue', 'borderwidth':5, 'relief':tk.RIDGE }

# This label is the only reference label in which we are going to match other label from this one.
Normallabel = tk.Label(root, text="Normal label content", cnf=label_config)
Normallabel.grid(row=0, column=1,padx=5,pady=5)

# This is the padx label which is going to add some space to the left and right to the text which is inside this label.
# Note: we didn't give any height and width to this label so we can see the effect of the padx parameter.
padxlabel= tk.Label(root, text="This label is for the padx.",cnf=label_config,padx=50)
padxlabel.grid(row=0,column=2,padx=5,pady=5)

# This is the pady label which is going to add some space to the top and bottom to the text which is inside this label.
# Note: we didn't give any height and width to this label so we can see the effect of the pady parameter.
padylabel=tk.Label(root, text="This label is for the pady",cnf=label_config,pady=50)
padylabel.grid(row=2, column=1,padx=5,pady=5)


#=======================================Justify content================================================

# THis is the justifu section in which we will
# now we need a long string to display the justify text

justify_string = "Hello this is manish kumar.(Justify=tk.LEFT)\n In this tutorial we are seeing the\
\n difference of justify, padx, pady, text, textvariable, and \
\n also we are going to see the difference b/w them and also we\
\n we will see that how to use them" 

justify_text = tk.StringVar()
justify_text.set(justify_string)
justifylabel =  tk.Label(root, textvariable=justify_text, cnf=label_config, height=10, width=50, justify=tk.LEFT)
justifylabel.grid(row=3, column=1,padx=5,pady=5)


# multi line string for the justify parameter 
justify_string = "Hello this is manish kumar.(Justify=tk.RIGHT)\n In this tutorial we are seeing the\
\n difference of justify, padx, pady, text, textvariable, and \
\n also we are going to see the difference b/w them and also we\
\n we will see that how to use them" 

justify_text = tk.StringVar()
justify_text.set(justify_string)
justifylabel =  tk.Label(root, textvariable=justify_text, cnf=label_config, height=10, width=50, justify=tk.RIGHT)
justifylabel.grid(row=3, column=2,padx=5,pady=5)


root.mainloop()
```

### 3.acitvebackgroun, activeforeground, bitmap, state

<p style="font-family:serif; font-size:20px;">Now we will move to next tutorial which is activebackground, activeforeground and bitmap <br> <br> In this we will see the effect of the activebackground and activeforeground we know this that background is for the color of the label and the foreground means the color of the text. So, we will se the effect when the cursor comes on the label. </p>

<ul style="font-family:serif; font-size:20px;" > 
    <li> <h6>activebackground:</h6> activebackground defines the background color to be displayed when the mouse is over the widget.</li> 
     <li><h6>activeforeground:</h6> activeforeground defines the background color to be dispolayed when the mouse is over the widget.</li>
    <li><h6>Bitmap:</h6> Bitmap is the attribute which unable us to display an image or a small size of graphic on the label or the widget, we can also use the bitmap to display the icons on the labels.</li>
    <li><h6>state:</h6> There are three state of any widget disabled, normal and active, the state defiens that the widget is responsive or not to some event. </li>
</ul>

```python
root = tk.Tk()

root.title("acivebackground , activeforground, bitmap")
root.geometry("600x300")

# now we need to make three labels for each attribute 
label_config = {'bg':"white",'fg':'Blue', 'borderwidth':5, 'relief':tk.RIDGE, 'height':10, 'width':20 }

# here widget state is set as tk.ACTIVE in both activeactivebackground_label AND activactiveforeground_label.
# you need to make set the state as tk.NORMAL to see the what effect will be applied when there is no activactivebackground
# or activactiveforeground.

activebackground_label = tk.Label(root, text="state = tk.ACTIVE\nactivebackground", cnf=label_config , activebackground='#78d6ff',
                                  state=tk.ACTIVE,)
activebackground_label.grid(row=0, column=1, padx=5, pady=5)

activeforeground_label = tk.Label(root, text= "state = tk.ACTIVE\nactiveforeground" , cnf=label_config, activeforeground='red',
                                 state= tk.ACTIVE, activebackground='white')
activeforeground_label.grid(row=0, column=2, padx=5, pady=5)



# bitmap section to display the bitmap image to the label 
logo = tk.PhotoImage(file="bitmap icons/user.png")
bitmap_label = tk.Label(root, image=logo, compound=tk.RIGHT, text="This is the bitmap icon:", bg="white")
bitmap_label.grid(row=1, column=0, padx=5, pady=5)


# button widget to see the effect of the activebackground when the event happens.
button = tk.Button(root, text="hello",activebackground='#78d6ff')
button.grid(row=2,column=1)

root.mainloop()
```
