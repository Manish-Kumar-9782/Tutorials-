## Menu widget


"Drop-down" menus are popular way to present the user with a number of choices, yet take up minumal space on the sface of the application when the user is not making a choice.

* A menubutton is the part that always appears on the application.
* A menu is the list of choice that appears only after the user clicks on the menubutton.
* To select a choice, the user can drag the muse from the menubutton down onto one of the cohices. Alternatively, they can click and release the menubutton: the chices will appear and stay until the uer clicks one pf them..
* The Unox version of Tkinter supports "tear-off menu" if you as the designer wish it, a dotted line will appear above the choice. The user can click on this line to "tear off" the menu: a new, seperate, independent window appears containing the choices. 

###### A choice displayed on a menu amy be any of these things:

* A simple command: a text string (or image) that the user can select to perform some operation.
* A cascade: a text string or image that the user can select to how anoter whole menu of choices.
* A checkbutto
* A group of radiobuttons

To create a menu widget, you have to create a Menubutton first.
```ruby
Mb = tk.Menubutton(master , option....)
Menu = tk.Menu(Mb, option.....) 
```

## Creating a simple menu

````python
# creating a root winodow

root = tk.Tk()

root.title("Menu Tutorial")
root.geometry("400x300")
root.iconbitmap("./bitmap icons/home.ico")
#-------------------------------------------------------------------------------------------------------------------

    # Now we will make Menubutton and the menu widget  
    
# First of all we will create a Menubutton, Menubutton used to create the menu
MB =  tk.Menubutton(root, text="Menu" , activebackground = '#eaa6f5', activeforeground = 'white')
MB.pack(pady=10)

# Now, we will make the menu, The master of the menu will be MB 

MENU = tk.Menu(MB , activebackground = "#eb8263", tearoff =0 , bg ='white')


# now we need to configure this menu inside the menubutton

MB.config(menu = MENU) 

""" Options
    menu : menu option is available in the Menubutton widget. It is used to associate the menubutton with a set of choices,
           set this option to the Menu object containing those choices. That menu object musst hav been created bu passing the 
           associated menubutton to the constructor as its first argument.  
"""

# now we need to add some item inside it, carefull every item contain a command option to perform some task when the option is selected
# For now we will add some commands to perform the some task 

MENU.add_command(label = "Option 1")
MENU.add_command(label = "Option 2")
MENU.add_command(label = "Option 3")
MENU.add_command(label = "Option 4")

# Note: This is the simple menu this does not do anything when any of the option is selected
# To perform an action when an optin is selected, we need to add some command inside the every command.
    
#-------------------------------------------------------------------------------------------------------------------

root.mainloop()
````

#### Creating Next Choice Menu with
* add_command
* add_cascade
* add_checkbutton
* add_Radiobutton

````python
# creating a root winodow

root = tk.Tk()

root.title("Menu Tutorial")
root.geometry("400x300")
root.iconbitmap("./bitmap icons/home.ico")
#-------------------------------------------------------------------------------------------------------------------

    # Now we will make Menubutton and the menu widget  
    
# First of all we will create a Menubutton, Menubutton used to create the menu
MB =  tk.Menubutton(root, text="Cascade Menu" , activebackground = '#eaa6f5', activeforeground = 'white')
MB.pack(pady=10)

# Now, we will make the menu, The master of the menu will be MB 

# all the menu which are inside the Menubutton should use the MB as the master 
# Even they are used in the cascade (Next choice menu) Menu

cascade_menu = tk.Menu(MB , activebackground = "#eb8263", tearoff =0 , bg ='white')
# Note: here only we will config the cascade menu to the Menubutton
# but remaining menus are configured inside the cascade menu to make the next choice menu

command_menu = tk.Menu(MB , activebackground = "#eb8263", tearoff =0 , bg ='white')
# command_menu will contain the commnand option as choice

checkbutton_menu = tk.Menu(MB , activebackground = "#eb8263", tearoff =0 , bg ='white')
# checkbutton_menu will contain the checkbuttons options as choice

Radiobutton_menu = tk.Menu(MB , activebackground = "#eb8263", tearoff =0 , bg ='white')
# radiobutton_menu  will contain the radiobuttons options as choice


# This list is for the next choice menu , they are fake option and dont call any callback function and also dont return any things
menu_options = ["option 1", "option 2", "option 3 "]

# NextMenu contains the labels of next choice menu inside the Main menu (cascade) 
NextMenu = ["Command Menu", "Checkbutton Menu", "Radiobutton Menu"]

# Next_Menu contains the menu object like command, checkbutton and radiobutton
Next_Menu = [command_menu, checkbutton_menu, Radiobutton_menu]


# Now we will only configure the cascade_menu to the MB widget.
MB.config(menu = cascade_menu) 

# NOte: all the other next choice menus are configured inside the cascade menu.

# now we need to add some item inside cascade menu,
# Every cascade can have next choice of menu 

# Now we will add
# command, checkbutton , radiobutton menu inside the cascade menus.

# first cascasde will contains the command options
# Second cascasde will contains the Checkbutton options
# Third cascasde will contains the Radiocheck options

    
for option in menu_options:
    command_menu.add_command(label = option)
    checkbutton_menu.add_checkbutton(label = option)
    Radiobutton_menu.add_radiobutton(label = option)
    
# Now we have put added the all the option inside to their desired location.
# Now we will Add 

for option  , MENU in zip(NextMenu , Next_Menu):
    cascade_menu.add_cascade(label=option , menu = MENU)

"""
    Result: In this we have created a 'cascade menu' as the main menu which will be displayed by clicking on the Menubutton.
            This cascade menu will contains some other next choice menus which are configured in the respective cascade menu.
            
            All the cascade menus contains:
            
            Command Menu: This contains the commads list which can perform some operations.
            
            Checkbutton Menu: This contains a list of options which is nothing but the checkbutoon options.
                              This menu will return Mulutiple selectd option values.
            
            Radiobutton Menu: This contains a list of options which is nothing but the radiobuttons options.
                              This menu will only a single selected value.

"""
print(f"This option is a type of {cascade_menu.type(1)} ")
print(f"This option is a type of {command_menu.type(1)} ")
print(f"This option is a type of {checkbutton_menu.type(1)} ")
print(f"This option is a type of {Radiobutton_menu.type(1)} ")
#-------------------------------------------------------------------------------------------------------------------
print(f"top level window  {MB.winfo_toplevel()}")

root.mainloop()
````
