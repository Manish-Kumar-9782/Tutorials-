### While Loop

```python
while expression:
    # note expression must give boolean result as true and false.
    break
```

````python
a = 1
while a<10:
    if a == 6:
        a+=1 # increasing a by 1.
        continue # skipping all the statement below this line.
    print("a is : ", a) 
    a += 1 # increasing a by 1.
````

```python
for var in range(0, 10):
    print(var)
```

````python
for var in range(0, 10):
    print(var, end=" ")
````
