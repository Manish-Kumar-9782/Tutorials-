```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import scipy
from scipy.interpolate import griddata
from scipy.optimize import curve_fit
```

```python
data = np.loadtxt('Gravity_UTM.txt')
utm_x = np.array(data[:,0])
utm_y = np.array(data[:,1])
CBA = np.array(data[:,2])
     
```

```python
utm_x
```

```python
xi = np.linspace(min(utm_x), max(utm_x), 300)
yi = np.linspace(min(utm_y), max(utm_y), 300)
xi,yi = np.meshgrid(xi,yi)
```

```python
# interpolation
zi = griddata((utm_x, utm_y), CBA, (xi,yi))
```

```python
xmin, xmax = min(utm_x), max(utm_x)
ymin, ymax = min(utm_x), max(utm_y)

plt.figure(figsize = (8,7))
plt.imshow(zi, aspect='auto', extent=(xmin, xmax, ymin, ymax), cmap="jet")
plt.colorbar()
```

**Cubic interpolation**

```python
# interpolation
zi = griddata((utm_x, utm_y), CBA, (xi,yi), method='cubic', rescale=True)

plt.figure(figsize = (10,7))

plt.contourf(xi,yi,zi, levels=75, cmap='jet')
plt.contour(xi,yi,zi, levels=75, linewidths=0.5, alpha=0.2, colors='k', antialiased =True)

plt.savefig("test.png")
```
