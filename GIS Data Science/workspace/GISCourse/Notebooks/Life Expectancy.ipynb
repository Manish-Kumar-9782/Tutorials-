```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
c_code_file = "../data/Countries Codes.csv"
life_file  = "../data/Life Expectancy Data.csv"
countries_file = "../data/ne_10m_admin_0_countries/ne_10m_admin_0_countries.shp"
```

```python
life_df = pd.read_csv(life_file)
code_df = pd.read_csv(c_code_file)
countries_gdf = gpd.read_file(countries_file)
```

```python
life_df.columns = life_df.columns.str.strip()
# lif_df.head()
```

**Inspecting Countries data**

```python
countries_gdf.head()
```

```python
# Total columns
cols = list(countries_gdf.columns)
cols.sort()
print(cols)
```

```python
countries_selected_columns = ["NAME", "ADMIN", "ADM0_A3", "SOVEREIGNT", 'SOV_A3', "SUBREGION", "ISO_A3", "geometry"]
```

```python
countries_gdf2 =  countries_gdf[countries_selected_columns]
countries_gdf2.head()
```

```python
rep_points =  countries_gdf2.representative_point()
rep_points
```

```python
countries_gdf2.loc[:,"LONG"], countries_gdf2.loc[:,"LAT"] = rep_points.x, rep_points.y 
```

```python
countries_gdf2
```

```python
countries_lat_long = countries_gdf2[["NAME","ISO_A3", "LONG","LAT"]]
```

```python
countries_lat_long
```

**Renaming the column for merge operation**

```python
countries_lat_long.rename(columns={"NAME":"Country"}, inplace=True)
```

```python
countries_lat_long
```

```python
life_df2 =  life_df.merge(countries_lat_long, on="Country")
life_df2.head()
```

**Visulalizing the data points**

```python
life_df2_year_group = life_df2.groupby("Year")
```

```python
life_df2_year_group.get_group(2015)
```

```python
life_df2_year_group.get_group(2015).columns
```

```python
fig, ax = plt.subplots(figsize= (15,15))

year = 2015
life_data = life_df2_year_group.get_group(year)

countries_gdf2.plot(ax=ax, color='gray')
ax.grid(True, lw=1, ls='--', alpha=0.3)

display_by = life_data["GDP"]

scatter = ax.scatter(life_data["LONG"], life_data["LAT"], s=display_by/100, c=display_by)
colorbar = plt.colorbar(scatter, label="Life expectancy Intensity", shrink=0.4)

handles, labels = scatter.legend_elements(prop='sizes', alpha=0.7)
legend1 = ax.legend(handles, labels)


ax.text(80.22, 21.78, "India", size="xx-small", color='black')
plt.show()
```

## Housing data set



```python
house_df = pd.read_csv("../data/housing.csv")
house_df.head()
```

```python
fig, axes = plt.subplots(figsize=(10,5))

proximity, index =  house_df.ocean_proximity.factorize()

scatter =  axes.scatter(house_df.longitude, house_df.latitude, 
                        c=proximity, alpha=0.4, cmap='jet', 
                        s=house_df.median_house_value/10000)

element, _ = scatter.legend_elements()


plt.legend(element, index)

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
data =  pd.read_csv("../data/45H_Final_Cu_Pb_Zn data.csv").dropna()

# Combine mineral concentrations (sum or average)
data['combined'] = data[['Cu', 'Zn', 'Pb']].sum(axis=1)

# Create a pivot table for heatmap
heatmap_data = data.pivot_table(index='Latitude', columns='Longitude', values='combined', fill_value=0)

# Plot heatmap
plt.figure(figsize=(10, 8))
# plt.imshow(heatmap_data, cmap='jet', aspect='auto',interpolation="spline16")
plt.contourf()
plt.colorbar(label='Combined Mineral Concentration')
plt.title('Combined Mineral Concentration Heatmap')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
# plt.xticks(ticks=np.arange(len(heatmap_data.columns)), labels=heatmap_data.columns, rotation=45)
# plt.yticks(ticks=np.arange(len(heatmap_data.index)), labels=heatmap_data.index)
plt.show()
```
