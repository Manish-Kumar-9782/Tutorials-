```python
import geopandas as gpd
import geodatasets as geodata
```

```python
world_filepath =  geodata.get_path("naturalearth.land")
world = gpd.read_file(world_filepath)
world.head()
```

```python
world_loan = gpd.read_file("kiva_loans/kiva_loans.shp")
```

```python
world_loan.head()
```

```python
ax = world.plot(figsize = (20,20), color="whitesmoke", edgecolor="black", linestyle=":")
world_loan.plot(ax=ax, color='green', markersize=2)
```
