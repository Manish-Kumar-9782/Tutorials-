Now we will three more GeoDataFrames, containing compsite locations (Point), foot trails (LineString) and county boundaries (Polygon).

```python
location_file_path = "Decptsofinterest/Decptsofinterest.shp"
foot_trails_file_path  = "DEC_roadstrails/Decroadstrails.shp"
counties_file_path = "NY_county_boundaries/NY_county_boundaries.shp"
```

```python
# point of interest data
poi_data = gpd.read_file(location_file_path)
poi_data.head()
```

```python
# Now from this dataset we only need campsit data points.
campsite = poi_data.loc[poi_data.ASSET=="PRIMITIVE CAMPSITE"].copy()
campsite.head()
```

**Plotting all campsite data point**

```python
fig, ax = plt.subplots()
# data.plot(ax = ax, color='orange')
wild_lands.plot(ax=ax, color='blue')
ax.plot(campsite.geometry.x, campsite.geometry.y, 'r.', markersize=3)
```

**Loading Foot Trails in new York State**

```python
foot_trails = gpd.read_file(foot_trails_file_path)
foot_trails.head()
```

```python
# extracting only ASSET by FOOT TRAIL
trails_on_foot = foot_trails.loc[foot_trails.ASSET == "FOOT TRAIL"].copy()
trails_on_foot.plot()
```

```python
# County boundries in New york state
counties = gpd.read_file(counties_file_path)
counties.head()
```

```python
# Define a base map with county boundaries
ax = counties.plot(figsize=(10,10), color='none', edgecolor='gainsboro', zorder=3)
wild_lands.plot(ax=ax, color='lightgreen')
campsite.plot(ax=ax, color='maroon', markersize=2)
trails_on_foot.plot(ax=ax, color='black', markersize=1)
```
