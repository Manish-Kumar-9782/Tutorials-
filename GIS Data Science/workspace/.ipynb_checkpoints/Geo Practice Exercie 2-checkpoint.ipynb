### Geographic coordinate sanity check

A **geographic coordinate sanity check** is a way to ensure that the coordinates (latitude and longitude) you are using or analyzing make sense and are accurate. Here’s a simple explanation:

- **Verify Validity**: Check if the coordinates fall within the valid range (latitude: -90 to 90, longitude: -180 to 180).
- **Check Location**: Ensure the coordinates correspond to a real-world location (e.g., not in the middle of the ocean if you expect a city).
- **Consistency**: Confirm that the coordinates are consistent with other data points or known locations.
- **Projection Accuracy**: Make sure the coordinates are correctly transformed if using different map projections or coordinate systems.

**Lenght of degree of longitude**

$$\text{Length of a degree of longitude} = \frac{\pi \cdot R \cdot \cos(\phi)}{180}$$

**Calculate:**

1. ✍️What is the length (in km) of a degree of longitude at:

    * 0° latitude (equator)
    * 90° latitude (pole)
    * 35° latitude? 49° latitude?

```python
# a function to calculate the length of longitude degree in km
def length_of_longitude_degree(latitude):
    R = 6371  # Earth's radius in km
    # since length of the longitude degree depends on the latitude.
    # Longitude will change as the latitude changes.
    return np.pi * R * np.cos(np.radians(latitude)) / 180
```

```python

#  Constants

# Function to calculate the length of a degree of longitude
def length_of_longitude_degree(latitude):
    R = 6371  # Earth's radius in km : Constants
    return np.pi * R * np.cos(np.radians(latitude)) / 180

# Generate latitudes from 0 to 90 degrees
latitudes = np.linspace(0, 90, 100)
lengths = length_of_longitude_degree(latitudes)

# Plotting
plt.figure(figsize=(6, 4))
plt.plot(latitudes, lengths, label='Length of Longitude Degree')
plt.scatter([35, 49], length_of_longitude_degree([35, 49]), color='red', zorder=5)
plt.text(35, length_of_longitude_degree(35), '  35°', color='red')
plt.text(49, length_of_longitude_degree(49), '  49°', color='red')
plt.xlabel('Latitude (degrees)')
plt.ylabel('Length of Longitude Degree (km)')
plt.title('Length of a Degree of Longitude as a Function of Latitude')
plt.legend()
plt.grid(True)
plt.show()
```

##### Let's Break down above code

To Calcuate the length of a degree of longitude at a give latitude, we can use the following formula.

$$\text{Length of a degree of longitude} = \frac{\pi \cdot R \cdot \cos(\phi)}{180}$$

In this formula we need to following compoenent:

1. $\pi$ : first we need value which is constant, for this we will use `np.pi` constant.
2. $R$ : Now we also need the radius of the Earth. (approximately `6371 km`)
3. $\phi$ : Now we need the latitude in degree for which we need to compute the longitue value.

**earth radius**

```python
R = 6371 # in km.
```

**function for lenght of longitude**

```python
def length_of_longitude_degree(lats):
    # we are using numpy for performing computation on multiple latitude value
    # as once.

    # Note np.cos functions takes values in radians se first we need to convert
    # latitude degree value into radians values.
    lat_rads = np.radians(lats)
    return (np.pi * R * np.cos(lat_rads))/180
```

**converting various latitude (degree) values**

`Note:` In this we are converting latitude in degree to longitude values in kilometer. 

```python
# testing our function
lats =  np.arange(0,90,10)
lon_lengths =  length_of_longitude_degree(lats)
# rounding values.
lon_lengths = np.round(lon_lengths,2)
lon_lengths
```

**key Observation**

* **Specific points:** The chart highlights two specific latitudes:

**Using Custom AEA projection whose origin is the center of glas_gdf bounds**

```python
x_min, y_min, x_max, y_max =  glas_gdf.total_bounds
x_center = (x_min + (x_max))/2
y_center = (y_min + (y_max))/2
print(x_center, y_center)
```

```python
utm_proj = f"+proj=aea +lat_1=33.33 +lat_2=46.67 +lat_0={y_center} +lon_0={x_center} \
+x_0=0 +y_0=0 +datum=WGS84 +units=m +no_defs"
```

```python
# Define the custom AEA projection
custom_aea = pyproj.CRS.from_proj4(utm_proj)


# Reproject the GeoDataFrame to the custom AEA projection
glas_gdf_aea = glas_gdf.to_crs(custom_aea)

# Plot the reprojected GeoDataFrame
glas_gdf_aea.plot(column="glas_z", markersize = 1)
plt.title('World Map in Custom Albers Equal Area Projection')
plt.axhline(0, lw=0.5, color='red')
plt.axvline(0, lw=0.5, color= 'red')
plt.xlabel('Easting (meters)')
plt.ylabel('Northing (meters)')
plt.show()
```

##### Calculating distance (Euclidian distance) for remaining projection.

**Euclidean Distance**: This is the straight-line distance between two points in a flat, two-dimensional plane. It does not consider the Earth’s curvature and is typically used for short distances or within small areas where curvature effects are negligible.

Since we are focusing on the samml region so curvature will not effect our calculation for the following remaining projectin. All of these follwowing projection has units in `meters`, so calculation will be also in meters.


1. UTM Zone 11N
2. Custom AEA Projection
3. Web Mercator

Steps:

1. First select the GeoDataFrame of a Projection
2. Azimuth Angle
    1. Use the same point for distance calculation.
    2. Calculate the difference in x and y direction ($dx$ and $dy$).
    3. Calculate the azimuth angle $\tan{\theta}$ in b/w ($dx$ and $dy$).
3. Distance
    1. Simply calculate the distance b/w points.
