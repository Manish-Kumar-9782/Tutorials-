**bounds:**  Returns a tuple containing the minimum and maximum x and y coordinates of the GeoDataFrame.

```python
# bounds: minimum and maximum x and y coordinates of the GeoDataFrame
muncipality.bounds
```

**total_bounds:**  Returns a tuple containing the minimum and maximum x and y coordinates of the GeoDataFrame, including any holes or gaps.

```python
muncipality.total_bounds
```

```python
fig, ax = plt.subplots()
muncipality.plot(ax = ax, color='white', edgecolor='black')
total_bounds = muncipality.total_bounds

ax.plot([total_bounds[0]], [total_bounds[1]], 'r.', linewidth=2) 
ax.plot([total_bounds[2]], [total_bounds[3]], 'g.', linewidth=2) 

# ax.plot([total_bounds[0], total_bounds[2]], [total_bounds[3], total_bounds[3]], 'g.', linewidth=2)
# ax.plot([total_bounds[0], total_bounds[0]], [total_bounds[1], total_bounds[3]], 'k-', linewidth=2)
# ax.plot([total_bounds[2], total_bounds[2]], [total_bounds[1], total_bounds[3]], 'k-', linewidth=2)
```
