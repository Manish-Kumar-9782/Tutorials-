## handling missing or null values 

```python
df
```

```python
df.isnull().sum()
```

```python
df_null =  df[["Country","Alcohol", "Hepatitis B", "Total expenditure", "GDP", "Population","Schooling", "Income composition of resources"]].copy()
```

```python
null_values_index = df_null[df_null.Alcohol.isnull()].index
df_null[df_null.Alcohol.isnull()]
```

```python
null_values_index
```

```python
df_null.iloc[null_values_index]
```

```python
df_null.fillna(method="bfill").iloc[null_values_index]
```

```python
df_null2 = df_null.copy()
df_null2.fillna(method="bfill", inplace=True)
df_null2.iloc[null_values_index]
```

```python
df_null2.iloc[30:40]
```

```python
df_null2.loc[32:34,"Alcohol"] = np.nan
df_null2.loc[37,"GDP"] = np.nan
df_null2.loc[35,"Schooling"] = np.nan
```

```python
df_null2[30:40]
```

```python
df_null2.fillna(method="ffill", axis=1).iloc[30:40]
# df_null2.iloc[30:40]
```
