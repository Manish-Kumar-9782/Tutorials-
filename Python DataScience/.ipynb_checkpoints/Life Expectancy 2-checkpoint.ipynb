```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("datasets/Life-Expectancy-Data-Updated.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df_2  = df[df.columns[3:]]
```

```python
df_2.head()
```

```python
corr = df_2.corr()
```

```python
corr
```

```python
fig, axes = plt.subplots(figsize=(15,10))
sns.heatmap(corr, annot = True, ax=axes)
plt.show()
```

## Normal Distribution

```python
fig, axes = plt.subplots(figsize=(15,6), ncols=6, nrows=3, tight_layout=True)

for col, _axes_ in zip(df_2.columns, axes.flatten()):
    sns.kdeplot(df_2[col], ax=_axes_)

plt.plot()
```

# Normal distribution

```python
fig, axes = plt.subplots(figsize = (15, 6))


rand_norm =  np.random.normal(50, 5, 3000)
sns.kdeplot(rand_norm, ax=axes)

axes.vlines([45,55], 0,.09)
```
