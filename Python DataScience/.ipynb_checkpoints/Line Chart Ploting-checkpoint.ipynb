## Ploting

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./datasets/pizza_delivery.csv")
```

```python
df.head()
```

```python
# df.sort_values("time", inplace=True)
```

```python
plt.plot(df.time, df.temperature)
```

Now sorting the data

```python
df.sort_values("time", inplace=True)
```

```python
plt.plot(df.time, df.temperature)
```

```python
df_sample = df.sample(frac=.3).sort_values("time")
```

```python
plt.plot(df_sample.time, df_sample.temperature)
plt.xlabel("Time Of Delivery")
plt.ylabel("Temperature on Delivery")
plt.title("Pizza Temperature vs Devliry Time")
```
