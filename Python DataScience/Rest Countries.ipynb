```python
import requests
import json
```

```python
req =  requests.get("https://restcountries.com/v3.1/region/europe")
```

```python
content=  json.loads(req.content.decode())
```

```python
req2 = requests.get("https://restcountries.com/v3.1/all?fields=name,region")

content2 = json.loads(req2.content.decode())
```

```python
countries =  [{"name":value["name"]["common"], "region":value["region"]} for value in content2]
```

```python
import pandas as pd
```

```python
cdf =  pd.DataFrame(countries)
cdf
```

```python
df = pd.read_csv("./datasets/Life Expectancy Data.csv")
```

```python
df.head()
```

```python
df.Country.unique()
```

```python
cdf2 =  cdf[cdf["name"].isin(df.Country.unique())]
cdf2
```

```python
AllCountries = pd.Series(df.Country.unique())
AllCountries
```

```python
SelectedCountries =  AllCountries[AllCountries.isin(cdf2.name)]
SelectedCountries
```

```python
new_df =  df[df.Country.isin(SelectedCountries)]
new_df
```

```python
new_df.Country.apply(lambda x: cdf2.set_index("name").loc[x].item())
```

```python
new_df["Region"] = new_df.Country.apply(lambda x: cdf2.set_index("name").loc[x].item())
```
