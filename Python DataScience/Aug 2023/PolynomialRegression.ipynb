```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

equation:

linear equation: $$y  = mx + c$$

polynomial regression: $$ y = ax^n + c$$

```python
## 2x^2 + 3x + 5

limit = 200
x = np.random.rand(limit) + np.random.randint(0,30,limit)

y =  0.6*x**2 + x + 5 

y += np.random.randn(limit) + np.random.randint(-50,40,limit)

plt.scatter(x,y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
```

```python
train_x, test_x, train_y, test_y = train_test_split(x,y, test_size=.2)
```

```python
reg = LinearRegression()
```

```python
reg.fit(train_x.reshape(-1,1),train_y.reshape(-1,1))
```

```python
y_pred = reg.predict(test_x.reshape(-1,1))
```

```python
plt.scatter(x,y)
plt.plot(test_x, y_pred, color='red')
```

```python
r2_score(test_y, y_pred)
```

```python
poly = PolynomialFeatures()
```

```python
train_xx =  poly.fit_transform(train_x.reshape(-1,1))
test_xx = poly.fit_transform(test_x.reshape(-1,1))
```

```python
reg = LinearRegression()
```

```python
reg.fit(train_xx,train_y.reshape(-1,1))
```

```python
y_pred = reg.predict(test_xx)
```

```python
df = pd.DataFrame({"test_x": test_x, "y_pred":y_pred.flatten()})
```

```python
df = df.sort_values("test_x")
```

```python
df
```

```python
plt.scatter(x,y)
plt.plot(df.test_x, df.y_pred, color='red')
```

```python
r2_score(test_y, y_pred)
```
