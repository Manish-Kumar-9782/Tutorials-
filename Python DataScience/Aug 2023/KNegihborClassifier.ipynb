```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris, make_moons, make_blobs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.inspection import DecisionBoundaryDisplay
```

## 1. Iris Classification

#### loading dataset

```python
iris_data.keys()
```

```python
iris_data = load_iris()
data_x = iris_data["data"]
data_y = iris_data["target"]

iris_df = pd.DataFrame(np.c_[data_x, data_y], columns=iris_data["feature_names"]+["classes"])
iris_df.columns = iris_df.columns.str.removesuffix("(cm)").str.strip()
iris_df.head()
```

### creating and training model

```python
knn = KNeighborsClassifier(n_neighbors=15)
knn.fit(iris_df.iloc[:,:2], iris_df.classes)
```

```python
def scatter(df, xlabel, ylabel):
    fig, ax1 = plt.subplots(figsize=(6,4))

    ax1.scatter(df[xlabel], df[ylabel], c=data_y, cmap=plt.cm.viridis)
    
    ax1.set_xlabel(xlabel)
    ax1.set_ylabel(ylabel)
    
```

```python
scatter(iris_df, "sepal length", "sepal width")
ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(knn, iris_df.iloc[:, :2], ax=ax, alpha=0.5)
```

## 2. Moons Classification

### loading data

```python
data_x, data_y = make_moons(n_samples=300, noise=0.15)
```

```python
df = pd.DataFrame(np.c_[data_x, data_y], columns=["x1","x2","class"])
df.head()
```

### creating and traning model

```python
knn = KNeighborsClassifier(n_neighbors=11, weights="distance")
knn.fit(data_x, data_y)
```

```python
scatter(df, "x1","x2")
ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(knn, data_x, ax=ax, alpha=0.5)
```

## 3. Multiple Cluster classification

```python
data_x, data_y = make_blobs(n_samples=300, centers=4, cluster_std=1.8, random_state=1)
df = pd.DataFrame(np.c_[data_x, data_y], columns=["x1","x2","class"])
df.head()
```

```python
knn = KNeighborsClassifier(n_neighbors=11, weights='uniform')
knn.fit(data_x, data_y)
```

```python
scatter(df, "x1","x2")
ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(knn, data_x, ax=ax, alpha=0.5)
```
