## Loops in javaScript

In javaScript we have two type of loop

1. for-loop
2. while-loop


`for-loop:`
1. for loop start with `for` keyword.
2. use `()` after `for` keyword.
3. now we need to make the block of the `for-loop`.
4. for-loop `()` contains three expressions.
    1. Intialization of the counter of loop.
    2. Termination condition of the for loop.
    3. Increment of counter variable.
    
`Syntax:`

```js
    for(initialization; condition; increment){
        // statements.
        /*
        Note: if expression of condition is 'true', loop will be in running mode until condition
            result become 'false'.
        */
    }   
```

#### for-loop first example

````javascript
for(let i=0; i<10; i++){
    // incrementing the counter variable by one.
        // statements.
    console.log(`i ---> ${i}`)
    }
````

**Example 2 with increased step**

````javascript
for(let i=0; i<10; i+=3){
    // Increasing the counter variable by 3.
        // statements.
    console.log(`i ---> ${i}`)
    }
````

`while-loop:`

1. starts with `while` keyword.
2. use `()` after `while` keyword.
3. Now make the block for `while-loop` by using `{}` braces.
4. define the condition inside the `()` of while loop.

```js
    while(conditon){
        // statement
        
        /*
         while loop will be in running mode until our condtion does not 
         return `false` value.
        */
    }

    /*
        To make an infinite loop we just use `true` or `1` inside the 
        () at palce of condition.
    */
```

**infinite-loop**

````javascript
// if we are placing `true` or `1` at the position of condition
// it means it is a infinite loop.
while(true){
    console.log("running")
}
````

<b style="color:red;">Warning:</b> Do not use the infinite loop until you don't know that what are you doing.

if we use the while (infinite) loop at the wrong place then our page may not load or script may hang browser.

**Control on while loop**

1. We can make condition to terminate the while loop
2. if we are using the infinite loop then we can use `break` keyword to terminate the loop at a specific point.
3. if we are using the inifinite loop inside the function then we can also
use the `return` keyword if we want to get-out from the loop as well as the funciton. 

````javascript
function isPrime(number){
    
    let i=2;
    // infinite loop
    while(1){
    
    // console.log("i at -> ", i);
    
    if (i==number){
        // console.log(`${number} is prime number.`);
        // break; 
        return true
    }
    else{
        
        if(number%i == 0){
            // console.log(`${number} is not a prime number`);
            // break; 
            return false
        }
    }
    i++;
}}
````

````javascript
// using the return keyword to break the loop
isPrime(10)
````

````javascript
function isPrimeBreak(number){
    
    let i=2;
    // infinite loop
    while(1){
    
    // console.log("i at -> ", i);
    
    if (i==number){
        console.log(`${number} is prime number.`);
        break; 
        // return true
    }
    else{
        
        if(number%i == 0){
            console.log(`${number} is not a prime number`);
            break; 
            // return false
        }
    }
    i++;
}}
````

````javascript
isPrimeBreak(10)
````

````javascript
isPrimeBreak(37)
````
