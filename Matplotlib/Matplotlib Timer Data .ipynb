### Getting Week total Working Time

```python
def get_oneweek_working_data(weekdata):
    
    working_index = ["Subject", "Project"]
    working_data = {}
    for date, data in weekdata.items():
        data = data.set_index("Record Type")
        working_data[date] = data.loc[data.index.intersection(working_index)].set_index("Record Topic").loc[:,"Total Time"]
        
    working_data2 = {}
    for date, data in working_data.items():
        current_date = {}
        for key, value in data.items():
            current_date[key] = current_date.get(key, 0)+ value
        working_data2[date] = current_date
    
    return working_data2
#------------------------------------------------------------------------------------------
def total_week_working_time(week_WK):
    total_time = 0
    for date, data in week_WK.items():
        total_time += sum(data.values())
    
    return total_time
```

### Getting Week Daily Activity Time

```python
def get_oneweek_activity_time(weekdata):
    
    working_index = ["ActivityClasses"]
    working_data = {}
    for date, data in weekdata.items():
        data = data.set_index("Record Type")
        working_data[date] = data.loc[data.index.intersection(working_index)].set_index("Record Topic").loc[:,"Total Time"]
        
    working_data2 = {}
    for date, data in working_data.items():
        current_date = {}
        for key, value in data.items():
            current_date[key] = current_date.get(key, 0)+ value
        working_data2[date] = current_date
        
    return working_data2
#------------------------------------------------------------------------------------------
def week_activity_time(week_AT):
    total_time = {}
    
    for date, data in week_AT.items():
        for key, value in data.items():
            total_time[key] = total_time.get(key, 0)+ value
    
    return total_time

#------------------------------------------------------------------------------------------
def total_week_activity_time(week_AT):
    total_time = 0
    for date, data in week_AT.items():
        total_time += sum(data.values())
    
    return total_time
```

### Work Time Curve
In this we will show our Work Time data with date which will show the increament and decrement of the work time within date

```python
data = pd.read_csv("./data/Timer_Data.csv")
```

```python
data.head()
```

```python
data = data.loc[3:,:]
```

```python
data.head()
```

Now first we need to change our Date column in datetime object column

```python
data['Date'] = pd.to_datetime(data['Date'], format="%d-%b-%Y")
```

```python
data = data.sort_values("Date")
```

```python
# To show the work time first of all we need to make a Data pipe line to fetch the proper data
def get_all_worktime(database):
    """
    This method will be used to retrieve all the work time data.
    """
    
    all_worktime = {}
    
    data = database.set_index("Record Type")
    Index  = data.index.intersection(["Subject", "Project"])
    data = data.loc[Index].set_index("Date", drop=False).loc[:,"Total Time"]
    
    for date, time in data.items():
        all_worktime[date] = all_worktime.get(date, 0) + time 
    
    return all_worktime

```

```python
all_worktime = get_all_worktime(data)
```

```python
all_worktime = dict(sorted(all_worktime.items(), key=lambda kv:(kv[0], kv[1])))
```

```python
all_worktime
```

```python
min_date = data["Date"].min()
max_date = data["Date"].max()
```

```python
print(min_date)
print(max_date)
```

Now we will plot this data 

```python
fig, ax = plt.subplots(figsize=(12, 5), dpi=100)

def time_formatter(value, pos):
    value = value /3600
    return f"{value:0.02f} hr"

# mean working hour 
mean = sum(all_worktime.values())/len(all_worktime.values())

ax.plot(all_worktime.keys(), all_worktime.values(), '--', color='#c47445')
ax.set_title("Work Time Curve")
ax.yaxis.set_major_formatter(time_formatter)

# here to create a hlines we need to specify the xmin and xmax value in the form in which our 
# xaxis fromatter is set, in this case our xaxis set in the date time formatter
# so we need to pass the minimum date for the series and maximlum for the series.
ax.hlines(mean, color='red', xmin=min_date, xmax=max_date, linestyle='dashed', label='Average line')
plt.show()
```
