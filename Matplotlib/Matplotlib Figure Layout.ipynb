# Cutomize Figure Layout

* subplots(): Perhaps the primary function used to create figures and axes. it's also similar to matplotlib.pyplot.subplot(), but crates and places all axes on the figure at once. 

* GridSpec: Specifies the geometry of the grid that a subplot will be placed. The number of rows and number of columns of the grid need to be set. Optionally, the subplot layout parameters acan be tuned.

* SubplotSpec: Specifies the location of the subplot in the given GridSpec.

Importing the necessary module 

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
```

### 1. 2x2 Grid layout
In this example we will create a 2x2 grid using both subplots() and gridspec.

if we want a equally spaced axes in the figure then the subplot is the best thing to use.

```python
fig , axs = plt.subplots(nrows=2, ncols=2, constrained_layout=True)
# nrow: number of rows in grid
# ncols: number of columns in grid
# constrained_layout: if True, it will adjust the elements of the figure.
```

Now this above figure layout can be achieved by using the `GridSpec` instance separately, then pass elements of gridspec instance to the add_subplot() metod to create the axes objects. The elements of the gridspec are accessed in generally the same manner as numpy arrays.

```python
# Now first of all we will create a figure with constrained_layout
fig = plt.figure(constrained_layout=True)

# Now we need to will make a grid of 2x2 with the help of the GridSpec.
spec = gridspec.GridSpec(ncols=2, nrows=2, figure=fig)
# Now this spec is the grid layout which can be accessed like an array element
# Now we will add a new subplot in the seleted spec
ax1 = fig.add_subplot(spec[0,0])
ax2 = fig.add_subplot(spec[0,1])
ax3 = fig.add_subplot(spec[1,0])
ax4 = fig.add_subplot(spec[1,1])
```

Now as you can see that we have achieved the same result as we have got with add_subplots() method.

### 2. Subplot with span rows & cols
The power of gridspec comes in being able to create subplots that span rows and columns.

```python
# First of all we will make a figure with constrained_layout
fig2 = plt.figure(constrained_layout=True)
# Now we will use the .add_gridspec() instead of directly using GridSpec
grid = fig2.add_gridspec(nrows=3, ncols=3)
# this will return a GridSpec instance which will having the 
# figure as fig itself.
# Now we will add some axes or subplot in these blocks or grids.
fig2_ax1 = fig2.add_subplot(grid[0, :])
fig2_ax1.set_title('grid[0, :]')

fig2_ax2 = fig2.add_subplot(grid[1,:-1])
fig2_ax2.set_title('grid[1, :-1]')

fig2_ax3 = fig2.add_subplot(grid[1:, -1])
fig2_ax3.set_title('grid[1:, -1]')

fig2_ax4 = fig2.add_subplot(grid[2, 0])
fig2_ax4.set_title('grid[2, 0]')

fig2_ax5 = fig2.add_subplot(grid[2, 1])
fig2_ax5.set_title('grid[2, 1]')

plt.show()
```

```python
# inspecting the grid
print(grid)
```

 So as we can see that both thing are same we can use both `gridspec.GridSpec` or `.add_gridspec` method to make a custom figure layout.
 
In above result we can see that we have created a more complex and spaned figure layout which can be usefull to show many time of data in one figure.

girdspec is alos essential for creating subplots of different width via a couple of methods.

### 3. Subplot with gridspec with different width and height

The method shown here is similar to the one above and initializes a uniform grid specification, and then uses numpy indexing and slices to allocate multiple "cells" for a given subplot.

```python
# for now will just create a uniform spec layout in which will put the subplot
fig = plt.figure(constrained_layout=True)

# Now we will make a grid layout
grid = fig.add_gridspec(nrows=2, ncols=2)

# in this we will use the .annotate method to put the text inside the subplot
anno_opt = dict(xy=(0.5, 0.5), xycoords='axes fraction',
               va='center', ha='center')

# Now we will add a subplot and we will put the annotation on it
fig.add_subplot(grid[0,0]).annotate('GridSpec[0,0]', **anno_opt)
fig.add_subplot(grid[0,1]).annotate('GridSpec[0,1]', **anno_opt)
fig.add_subplot(grid[1,0]).annotate('GridSpec[1,0]', **anno_opt)
fig.add_subplot(grid[1,1]).annotate('GridSpec[1,1]', **anno_opt)
plt.show()
```

Now we can modify this layout by using the another method in which  we can use width_ratio and height_ratio parameters. 

These keyword arguments are lists of numbers. Note that absolute values are meaningless, only their relative ratios matter. That means that width_ratios=[2, 4, 8] is equivalent to width_ratios=[1, 2, 4] within equally wide figures. 

For Example:

```python
%%time
fig = plt.figure(constrained_layout=True)
widths = [2, 3, 1.5]   # for width ratio
heights = [1, 3, 2]    # for height ratio

# Now we will add a gridspec in which we will use width_ratio and height_ratio
grid = fig.add_gridspec(nrows=3, ncols=3, width_ratios=widths,
                          height_ratios=heights)

# now we will use the for loop to create the subplot
for row in range(3):
    for col in range(3):
        ax = fig.add_subplot(grid[row, col])
        label = "widht:{}\nheight:{}".format(widths[col], heights[row])
        ax.annotate(label, (0.1, 0.5), xycoords='axes fraction', va='center')
        
plt.show()
```

### 4. width_ratios & height_ratios with subplots()
width_ratios and height_ratios can be particularly usefull  since the top-level function subplots() accepts them wtithn the gridspec_kw parameter. so we can pass any parameter which is accepted by GridSpec can be pased to subplots() via the gridspec_kw parameter.

Now we will recreate the previous figure without directly using the gridspec instance.

```python
%%time
widths = [2, 3, 1.5]   # for width ratio
heights = [1, 3, 2]    # for height ratio

gs_kw = dict(width_ratios=widths, height_ratios=heights)

fig, axs = plt.subplots(ncols=3, nrows=3, constrained_layout=True,
                      gridspec_kw=gs_kw)

for r, row in enumerate(axs):
    for c, ax in enumerate(row):
        label = 'Width: {}\nHeight: {}'.format(widths[c], heights[r])
        ax.annotate(label, (0.1, 0.5), xycoords='axes fraction', va='center')
plt.show()
```

### 5. Combining subplots & get_gridspec 
The subplots and get_gridspec methods can be combined since it is sometimes more convenient to make most of the subplots using subplots and then remove some and combine them. Here we create a layout with the bottom two axes in the last column combined.

```python
# first of all we will create a equally distributed grid layout
# by using the subplots

fig, axs = plt.subplots(ncols=3, nrows=3)
# Note: in this we are not using the constrained_layout parameter
# above line will create 3x3 grid layout

# Now we will get the gridspec
# Note: Note that grid spec is a 2D array so we can perform slicing 
# on it like we do on an array.
grid = axs[1, 2] # selecting the 2nd row and 3rd column
```

```python
# if we now inspect the gird
print(grid)
print(type(grid))
```

### 6. Fine Adjustments to a Gridspec Layout

When a GridSpec is explicitly used, you can adjust the layout parameters of subplots that are created from the GridSpec.

`Note:` This option is  not compatible with constrained_layout of figure.tight_layout which both adjust subplot size to fill the figure.

```python
fig = plt.figure(constrained_layout=False)
grid = fig.add_gridspec(ncols=3, nrows=3, left=0.05, right=0.48, wspace=0.05)
ax1 = fig.add_subplot(grid[:-1, :])  # selecting all columns and two rows
ax2 = fig.add_subplot(grid[-1, :-1])
ax3 = fig.add_subplot(grid[-1, -1])
plt.show()
```

his is similar to subplots_adjust(), but it only affects the subplots that are created from the given GridSpec.

For example, compare the left and right sides of this figure:

```python
fig = plt.figure(constrained_layout=False)
grid = fig.add_gridspec(ncols=3, nrows=3, left=0.05, right=0.48, wspace=0.15)
ax1 = fig.add_subplot(grid[:-1, :])  # selecting all columns and two rows
ax2 = fig.add_subplot(grid[-1, :-1])
ax3 = fig.add_subplot(grid[-1, -1])

# adding another gridspec to add more axes 
grid2 = fig.add_gridspec(ncols=3, nrows=3, left=0.55, right=0.98, hspace=0.20)

# adding more axes
ax4 = fig.add_subplot(grid2[:, :-1])  # selecting all row and two column
ax5 = fig.add_subplot(grid2[:-1, -1])
ax6 = fig.add_subplot(grid2[-1, -1])

plt.show()
```

### 7. GridSpec using SubplotSpec
You can create GridSpec from the SubplotSpec, in which case its layout parameters are set to that of the location of the given SubplotSpec.

Note this is also available from the more verbose gridspec.GridSpecFromSubplotSpec.

```python
%%time
fig = plt.figure(constrained_layout=True)

# Now we will make a grid with one row two column
grid = fig.add_gridspec(1,2)

# Now we will make subgridspec inside the our gird.
grid00 = grid[0].subgridspec(2,3)
grid01 = grid[1].subgridspec(3,2)

for i in range(2):
    for j in range(3):
        fig.add_subplot(grid00[i,j])
        fig.add_subplot(grid01[j,i])

plt.show()
```

```python
%%time
import numpy as np


def squiggle_xy(a, b, c, d, i=np.arange(0.0, 2*np.pi, 0.05)):
    return np.sin(i*a)*np.cos(i*b), np.sin(i*c)*np.cos(i*d)


fig11 = plt.figure(figsize=(8, 8), constrained_layout=False)
outer_grid = fig11.add_gridspec(4, 4, wspace=0, hspace=0)

for a in range(4):
    for b in range(4):
        # gridspec inside gridspec
        inner_grid = outer_grid[a, b].subgridspec(3, 3, wspace=0, hspace=0)
        axs = inner_grid.subplots()  # Create all subplots for the inner grid.
        for (c, d), ax in np.ndenumerate(axs):
            ax.plot(*squiggle_xy(a + 1, b + 1, c + 1, d + 1))
            ax.set(xticks=[], yticks=[])

# show only the outside spines
for ax in fig11.get_axes():
    ss = ax.get_subplotspec()
    ax.spines.get('top').set_visible(ss)
    ax.spines.get('bottom').set_visible(ss)
    ax.spines.get('left').set_visible(ss)
    ax.spines.get('right').set_visible(ss)

plt.show()
```
