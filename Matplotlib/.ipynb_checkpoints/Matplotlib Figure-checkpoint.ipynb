# Figure class
A Figure is a Top level artist, which holds all plot element. many methods are mplemented in FigureBase.

```ruby
matplotlib.figure.Figure(figsize=None, dpi=None, facecolor=None, edgecolor=None, linewidth=0.0, frameon=None, subplotpars=None, tight_layout=None, constrained_layout=None)[source]
```

 first of all we need to  import the matplotlib module.

````python
from matplotlib.figure import Figure
import matplotlib.pyplot as plt
import numpy as np
````

## 1. Creating a simple figure object

Note: we will use the pyplot to show the all the figure instance.

````python
# to create a figure which is a toplevel container for any kind of plot 
# so first of all we need to create a Figure object

fig = Figure(figsize=(7, 4), dpi=100)   # if we want
# instance of matplotlib.figure.Figure
Fig = plt.figure(figsize=(7, 4), dpi=100)
# instance of matplotlib.figure.Figure

# both gives us an instance of matplotlib.figure.Figure
# so we fig and Fig both instance share common attribute so we can apply on the method on Fig which can be applied on fig.

# Now after creating the figure object we need to add a subplot 
subplot = Fig.add_subplot()
subplot.plot([i for i in range(20)], [i**2 for i in range(20)])

# please read the matplotlib Figure documentation to know about the add_subplot arguments
# it takes three arguments nrow, ncol and index 
# which defines the position of the subplot in the Figure container and we can see that it uses the row and col to put the graph
# after adding the subplot it will return a Axes.SubplotBase object in which we can draw any kind of plot
plt.show()
````

````python
# we can add multiple plot in the same Figure like we have addd below
Fig = plt.figure(figsize=(7, 4), dpi=100)
subplot1 = Fig.add_subplot(541, position=[1,0,1,1])
subplot2 = Fig.add_subplot(542, position=[0,0,1,1])
subplot1.plot([i for i in range(20)], [i**2 for i in range(20)])
subplot2.plot([i for i in range(20)], [i**6 for i in range(20)])
plt.show()
````
