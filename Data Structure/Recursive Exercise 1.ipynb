## 1.Print a pattern without using any loop

"""
Input: n = 16

Output: 16, 11, 6, 1, -4, 1, 6, 11, 16

Input: n = 10

Output: 10, 5, 0, 5, 10
"""


```python
def pattern1(n,m,flag=False):
    print(n,end=" ")
    dif =5
    # if it cross the at once then it should not change the sign
    if n<=0:
        flag =True
        
    if n == m and flag == True :
        return
    
    if flag==False:
        
        return pattern1(n-dif,m,flag)
    else:
        
        return pattern1(n+dif,m,flag)
    
   
    
    return n
```

```python
pattern1(16,16)
```

```python
def printPattern(n):
    
    if n<=0:
        print(n,end=", ")
        return
    
    print(n, end=", ")  #
    printPattern(n-5 )  # These two line will excute untill it reaches to n<=0
    
    print(n, end=", ")  # after reaching at n<=0 this will strat to run in every call of the function 
    
```

## 2. Print all non-increasing sequences of sum equal to a given number x

Given a number x, print all possible non-increasing sequences with sum equals to x.
Examples: 

Input: x = 3

Output: 

1 1 1

2 1

3

Input: x = 4

Output:

1 1 1 1

2 1 1

2 2

3 1

4



## itrative method

```python
def Inc_pattern(n):
    
    for i in range(1,n+1):
        for j in range(0,i):
            print(i,end=" ")
        print()
        
Inc_pattern(3)
```

```python
def dec_pattern(n):
    
    for i in range(1, n+1):
        for j in range(n-i+1):
            print(i,end=" ")
        print()

dec_pattern(4)
```
