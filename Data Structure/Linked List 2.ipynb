## Linked List

### 1. Create a Node for Linked List

```python
# First we need to create a node class

class Node(object):
    def __init__(self, data):
        self.data = data # This will handle (store) the data
        self.next = None # this is a pointer which will refer to next Node
```

```python
class LinkedList(object):
    def __init__(self, head=None):
        self.head = head
        # This is the head of the first Node Which will refer to the very first
        # Node address

#=====================================================================================
    def print_Llist(self):
        # storeing the first head to the temp variable
        temp =self.head
        print()
        # Now we will check the Links of the LinkedList up to the end of 
        # the LinkedList
        while temp:
            print(temp.data)
            
            # temp.next Refers to the next node which saved in First Node
            # or next Node
            temp = temp.next
            
            # Here temp is a node which have following variable
            # temp.data
            # temp.next
            """
            try:
                print(f"\ntemp is a {temp}")
                print(f"type of the temp variable is {type(temp)}")
                print((f"type of the temp.data variable is {type(temp.data)}"))
                print((f"type of the temp.next variable is {type(temp.next)}\n"))
            except:
                print("\nLinkedList Node has ended up")
            """         
#=============================================================================                
    def get_position(self,position):
        pos = 0
        temp = self.head
        
        if self.head:
            while temp:
                pos += 1
                 # shift to next node
                if position == pos:
                    return temp.data
                temp = temp.next
        else:
            return None
        
        return None

#============================================================================  
    def insert(self,data,position):
        temp = self.head
        index = 0
        new_element = Node(data)
        # check the position if it exist:
        if temp:
            while temp!=None:
                index+=1
                
                
                if position == 1:
                    # inser at very first location only
                    self.head = new_element
                    # copy the remaining element
                    new_element.next = temp
                    break
                elif position-1 == index:
                    # at n-1 position
                    # copy the n-1 position element into last_element vaiable
                    last_element = temp.next # n position 'old element'
                    temp.next = new_element  # n 'position new element'
                    
                    # now link the last element to the new_element
                    new_element.next = last_element
                    break  
                
                temp = temp.next
                
        else:
            self.head = new_element

#=========================================================================
    def append(self,data):
        # insert an element at the end of the 
        temp = self.head
        index = 0
        if temp:
            #print(f"\nIn append  temp is {temp}")
            # Here, Since temp.next represt to the next Node
            # so we will check for the next Node 
            # if there is a next node it will be copied to temp
            # if there isn't no next node the loop will break 
            # Then we will access the last Node of the Linked list
            while temp.next:
                index+=1
                #print(f"temp at {index}  is {temp.data}")
                temp = temp.next
                
            # wait untill the last element comes into temp
            # assign new node to last element
            #print(f"\n\nlast element is at {index}")
            
            temp.next = Node(data)
        else:
            self.head = Node(data)
            
#=========================================================================

    def delete(self, position):
        """
        Algorithm: First we will make a copy/reference of head into temp
                   if temp exist:
                       go upto the delete position-1.
                       make two copy of linked List.
                       One for Privoius to delete and another for next to delete.
                       join the Privious and next one NOde and delete the old temp.
                   else:
                       make empty to head.
        """
        
        
        
        
        temp = self.head
        index = 0
        
        if temp:
            while temp:
                index += 1
                
                if position == 1:
                    self.head = temp.next
                    temp = None
                    break
                    
                elif position-1 == index:
                    # make a copy of previous  to a new temp1
                    # make a copy of last to new temp2
                    temp1 = temp
                    temp = temp.next # shifting to 
                    temp2 = temp.next
                    
                    # deleting the old Linked List
                    temp=None
                    # Making A new Linked List
                    temp1.next = temp2
                    break 
                temp = temp.next
        else:
            self.head = None
```
