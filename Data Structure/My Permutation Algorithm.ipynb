```python
# we need to make all possible combibation for any array

# for that we need an array 

Array  = [1,2,3]
n = len(Array)

# Now we nee to swap the elements in that way that they dont repeat their location/ Position

def permutation(a ,k, n ):
    ar = a
    # here a is the array which is going to be take part in permutation
    # k is the current element for swap
    # n is the total element 
    
    # our first combination is same as the input
    
    #print(f"k is {k}")
  

    # Keep in mind this condition we will see this conditions as an end of itration for every call
    # EOI = end of iteration 
    
    if k == 1:
        print(ar)
        return 
 
    for i in range(k):
        permutation(a,k-1,n )
        
        # this recursive method runs n! time means if you have 3 element in your array
        # it will run 3 X 2 X 1 = 6 time 
        # and every time it will return an combination of an array
        
        
        ar = a
        if k % 2 == 0:
            #print(f"k is {k} and {i}")
            ar[0], ar[k-1] = ar[k-1], ar[0]
        else:
            ar[i], ar[k-1] = ar[k-1], ar[i]
        
   
    
        
```
