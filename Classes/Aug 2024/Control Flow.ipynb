## Conditional Statement

#### only Single condition

```python
if condition:
    # statement
```

#### two Single condition

**yes or no (this or that)**

```python
if condition:
    # statement
else:
    # statement
```

#### multiple Single condition

**if-elif, elif, else**
```python
if condition:
    # statement
elif condition2:
    # statment
elif condition3:
    # statement
else:
    # finale statement
```

`Note:` returned value of condition for `if` or `elif` should be single boolean value.

````python
a = 7
b = 2
````

````python
if a<b:
    print("a is less than b")
else:
    print("a is greater than b")
````

````python
PASSWORD = "zxc123vbn"
# storing a password value in a variable
# later we will match this password with user input. (user password)
````

````python
OTP = random.randint(100,999)
# OTP : random.randint(100,999) will create a single random number in b/w the range.

print("otp number: ", OTP) # printing the OTP value generated by randomly.
# it is like sending an otp to the user/client

user_password = input("User Passwrod: ")
# taking user password using input function.
user_otp = input("Enter OTP: ")
# taking user otp which is sent to client.
# Note: note that our otp value must by int (integer) type 

user_otp = int(user_otp)  # converting our otp from str to int
# Now our user_otp contains int type value.

if user_password == PASSWORD and user_otp == OTP:
    print("Access Granted")
else:
    print("Access Denied")
````

## for-loop

```python
for <var> in <sequence>|<iterable>:
    # some statements
```

````python
for i in range(10):
    print("=>",i)
````

**Understanding range**

```python

    range(start)
    range(start, stop)
    range(start, stop, step)

    # start value is inclusive
    # stop value is exclusive

    # single value
    range(n)
    # start: 0, stop: n


    # start and stop value
    range(n1, n2)
    # start: n1, stop: n2

    # start, stop and step

    range(n1,n2,n3)
    # start: n1, stop: n2, step: n3
```.

````python
for i in range(10,20):
    print(i)
````

````python
for i in range(10,20,3):
    print(i)
````

````python
num = 7
for i in range(2,num):
    print(i," ", num,"%", i, " ",num%i==0)
````

Note: if we get `True` in for any test in range. it means our number is not prime number.

````python
num = 8
for i in range(2,num):
    print(i," ", num,"%", i, " ",num%i==0)
````

Note: if we get first `True` for any test in range, then we don't test any more. 

````python
num = 9
for i in range(2,num):
    print(i," ", num,"%", i, " ",num%i==0)
    if num%i == 0:
        # if we get True for any test then we will break (stop) the for loop
        break
````

**Summarized program**

````python
isPrime = True # initially we are considering that our number is prime number.
num = 7
for i in range(2,num):
    print(i," ", num,"%", i, " ",num%i==0)
    if num%i == 0:
        # if we get True for any test then we will break (stop) the for loop
        isPrime = False
        # since our number is divisible form another number.
        break

print("Number is Prime: ", isPrime)
````
